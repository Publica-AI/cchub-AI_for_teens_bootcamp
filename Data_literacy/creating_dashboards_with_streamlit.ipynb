# Section 1 — What is Streamlit? + Setup
### 15 minutes

## What is Streamlit?

Streamlit is a **Python library** that turns your script into a live website in seconds.

- You write Python code in a `.py` file
- You run one command in the terminal
- A website opens in your browser automatically — **no HTML, no CSS needed!**

Companies like data startups, MPESA analytics teams, and hospitals use tools like Streamlit to build internal dashboards fast.

## How to run any example in this notebook

1. **Copy** the code from the cell
2. **Paste** it into a new `.py` file using VS Code

4. A browser tab opens automatically. Press `Ctrl + C` to stop.

### Example 1 — Your First Streamlit App

1. Create a file called **`s1_hello.py`**
2. Copy the code below into it and save
3. Run: `streamlit run s1_hello.py`

````python
# FILE: s1_hello.py
# Terminal: streamlit run s1_hello.py

import streamlit as st

# set_page_config MUST be the first Streamlit command
st.set_page_config(page_title="Habari Nairobi", page_icon="K")

st.title("Habari, Nairobi!")
st.header("Welcome to my first Streamlit app")
st.write("My name is Kamau and I built this in Python. Sawa sawa!")

# Coloured message boxes
st.success("Streamlit is working!")
st.info("This is an info box")
st.warning("This is a warning box")
````

> **Try it:** Change `"My name is Kamau"` to your own name, save the file, and watch the browser update automatically!

# Section 2 — Displaying Data
### 20 minutes

The main job of a dashboard is to **show data clearly**.

| Tool | What it does |
|------|--------------|
| `st.metric(label, value)` | Big KPI number — like a scoreboard |
| `st.dataframe(df)` | Scrollable table |
| `st.bar_chart(data)` | Bar chart |
| `st.columns(n)` | Place things side by side |

You will need **pandas** to load and work with CSV files.

### Example 2 — Table, Chart, and Metric Cards

1. Create a file called **`s2_data_app.py`**
2. Copy the code below and save
3. Run: `streamlit run s2_data_app.py`

````python
# FILE: s2_data_app.py
# Terminal: streamlit run s2_data_app.py

import streamlit as st
import pandas as pd

st.set_page_config(page_title="Matunda Market", page_icon="O")
st.title("Matunda Market — Nairobi Prices")

# Build a small DataFrame
data = {
    "Fruit":     ["Mango",  "Avocado", "Banana", "Passion", "Guava"],
    "Price_KES": [25,        20,         10,       15,        18],
    "Stock_kg":  [50,        80,         120,      40,        60]
}
df = pd.DataFrame(data)

# Metric cards side by side
col1, col2, col3 = st.columns(3)
col1.metric("Fruits Listed",   len(df))
col2.metric("Cheapest (KES)",  df["Price_KES"].min())
col3.metric("Most Stock (kg)", df["Stock_kg"].max())

st.divider()

# Table
st.subheader("Price List")
st.dataframe(df, use_container_width=True)

# Bar chart
st.subheader("Stock Levels")
st.bar_chart(df.set_index("Fruit")["Stock_kg"])
````

# Section 3 — Making it Interactive
### 20 minutes

What makes a dashboard **powerful** is that users can filter the data themselves.

| Widget | Code | What it shows |
|--------|------|---------------|
| Dropdown | `st.selectbox("Label", options)` | Pick one item |
| Multi-select | `st.multiselect("Label", options)` | Pick many items |
| Slider | `st.slider("Label", min, max)` | Drag a number |

Every widget **returns a value** you can use to filter your DataFrame.

### Example 3 — Filter Data with a Sidebar

1. Create a file called **`s3_interactive.py`**
2. Copy the code below and save
3. Run: `streamlit run s3_interactive.py`

````python
# FILE: s3_interactive.py
# Terminal: streamlit run s3_interactive.py

import streamlit as st
import pandas as pd

st.set_page_config(page_title="Matatu Routes", page_icon="B")
st.title("Nairobi Matatu Tracker")

# Sample data
df = pd.DataFrame({
    "Route":    ["CBD - Westlands", "CBD - Karen", "CBD - Eastleigh",
                 "CBD - Kibera",    "CBD - Thika Rd"],
    "Fare_KES": [50, 80, 40, 30, 60],
    "Status":   ["On Time", "Delayed", "On Time", "On Time", "Delayed"]
})

# Sidebar: filter controls
with st.sidebar:
    st.header("Filter")

    # Dropdown — returns the selected string
    status_filter = st.selectbox("Show routes:", ["All", "On Time", "Delayed"])

    # Slider — returns the chosen number
    max_fare = st.slider("Max Fare (KES)", 20, 100, 80)

# Apply filters
filtered = df[df["Fare_KES"] <= max_fare]
if status_filter != "All":
    filtered = filtered[filtered["Status"] == status_filter]

# Show results
st.metric("Routes found", len(filtered))
st.dataframe(filtered, use_container_width=True)
st.bar_chart(filtered.set_index("Route")["Fare_KES"])
````

### What you should see
A sidebar on the left. Changing the dropdown or dragging the slider updates the table and chart instantly!

### Example 4 — The Student Dashboard

1. Create a file called **`s4_dashboard.py`** in the `Streamlit_Dashboards` folder
2. Copy the code below and save
3. Run: `streamlit run s4_dashboard.py`

````python
# FILE: s4_dashboard.py
# Terminal: streamlit run s4_dashboard.py

import streamlit as st
import pandas as pd

st.set_page_config(
    page_title="CCUB Bootcamp Dashboard",
    page_icon="G",
    layout="wide"
)

# @st.cache_data means the CSV is only read once, making the app faster
@st.cache_data
def load_data():
    return pd.read_csv("../Data_literacy/data/kenyan_student_data.csv")

df = load_data()

# Sidebar filters
with st.sidebar:
    st.title("Filters")

    all_tracks = sorted(df["Course_Track"].unique())
    selected_tracks = st.multiselect("Course Track", all_tracks, default=all_tracks)

    all_cities = sorted(df["City"].unique())
    selected_cities = st.multiselect("City", all_cities, default=all_cities)

# Apply filters
filtered = df[
    df["Course_Track"].isin(selected_tracks) &
    df["City"].isin(selected_cities)
]

# Page title
st.title("CCUB Bootcamp — Student Dashboard")
st.caption(f"Showing {len(filtered)} of {len(df)} students")
st.divider()

# 4 Metric cards
k1, k2, k3, k4 = st.columns(4)
k1.metric("Total Students",     len(filtered))
k2.metric("Cities",             filtered["City"].nunique())
k3.metric("PC Makes",           filtered["PC_make"].nunique())
k4.metric("Excellent Feedback", (filtered["Feedback"] == "Excellent").sum())

st.divider()

# 4 Charts in a 2x2 grid
c1, c2 = st.columns(2)
with c1:
    st.subheader("Students per Course Track")
    st.bar_chart(filtered["Course_Track"].value_counts())
with c2:
    st.subheader("Students per City")
    st.bar_chart(filtered["City"].value_counts())

c3, c4 = st.columns(2)
with c3:
    st.subheader("Gender Split")
    st.bar_chart(filtered["Gender"].value_counts())
with c4:
    st.subheader("Feedback Ratings")
    st.bar_chart(filtered["Feedback"].value_counts())

st.divider()

# Full student table
st.subheader("Student List")
st.dataframe(filtered.reset_index(drop=True), use_container_width=True)
````

### What you should see
- Sidebar filters — changing them updates **everything** on the page instantly
- 4 metric cards across the top
- 4 bar charts in a 2x2 grid
- Full student table at the bottom
