## 2. Loading and Saving Data

### Concept Explanation

In real life, data lives in **files** — not in your Python code. The most common file format for data is a **CSV** (Comma-Separated Values) file, which you can open in Excel or Google Sheets.

Pandas makes it easy to load data from a file and save results back to a file.

| Operation | Function | What it does |
|-----------|----------|--------------|
| Load CSV | `pd.read_csv("file.csv")` | Read a CSV file into a DataFrame |
| Save CSV | `df.to_csv("file.csv", index=False)` | Save a DataFrame to a CSV file |
| Load Excel | `pd.read_excel("file.xlsx")` | Read an Excel file |
| Preview first rows | `df.head(n)` | Show the first `n` rows (default 5) |

> 💡 `index=False` in `to_csv()` stops Pandas from writing the row numbers (0, 1, 2…) as an extra column in your file.

## 4. Data Cleaning — Handling Missing Values

### Concept Explanation

Real-world data is messy. Students miss an exam, a form is left blank, or a sensor fails to record — these gaps become **missing values** (shown as `NaN` — *Not a Number* — in Pandas).

Before analysing data, you must **clean** it by deciding what to do with these gaps.

> 🎯 **Analogy:** If the science exam register has a blank score for Odhiambo, you need to decide: was he absent? Should you put 0? Should you use the class average? Or should you remove his row entirely? That decision is data cleaning.

**Key missing-value functions:**

| Function | What it does |
|----------|--------------|
| `df.isnull()` | Returns True/False for every cell — True = missing |
| `df.isnull().sum()` | Counts missing values per column |
| `df.dropna()` | Removes ALL rows that have any missing value |
| `df.fillna(value)` | Fills missing values with a specified value |
| `df["col"].fillna(df["col"].mean())` | Fill with the column average |

````python
# Example 1: Find missing values

import pandas as pd

df = pd.read_csv("data/students.csv")

# Count missing values in each column
print("Missing values per column:")
print(df.isnull().sum())
print()

# What percentage of each column is missing?
missing_pct = (df.isnull().sum() / len(df)) * 100
print("Missing values (%):")
print(missing_pct.round(1))
````

````python
# Example 2: Fill missing values with the column mean
# Scenario: Students with missing science scores get the class average instead.

import pandas as pd

df = pd.read_csv("data/students.csv")

# Check who is missing science_score
print("Rows with missing science score:")
print(df[df["science_score"].isnull()][["name", "science_score"]])
print()

# Calculate the mean science score (ignoring NaN automatically)
mean_science = df["science_score"].mean()
print(f"Class average science score: {mean_science:.1f}")

# Fill the missing values with the mean
df["science_score"] = df["science_score"].fillna(mean_science)

print("\nAfter filling missing values:")
print(df[["name", "science_score"]])
````

````python
# Example 3: dropna() vs fillna() — when to use each

import pandas as pd

df = pd.read_csv("data/students.csv")

print(f"Original rows    : {len(df)}")

# dropna() removes any row with at least one missing value
df_dropped = df.dropna()
print(f"After dropna()   : {len(df_dropped)} rows (lost {len(df) - len(df_dropped)} rows)")

# fillna() keeps all rows — just replaces the gaps
df_filled = df.fillna(0)   # fill everything missing with 0
print(f"After fillna(0)  : {len(df_filled)} rows (no rows lost)")

print()
print("Rule of thumb:")
print("  Use dropna()  → when missing rows are few and won't skew results")
print("  Use fillna()  → when you can't afford to lose rows")
````

## 5. Accessing and Selecting Data

### Concept Explanation

Once your data is loaded, you need to **pick out** the specific parts you want — like finding one student's row, or pulling out just the scores column.

**Three main selection methods:**

| Method | Used for | Example |
|--------|----------|---------|
| `df["col"]` | Select one column | `df["name"]` |
| `df[["col1", "col2"]]` | Select multiple columns | `df[["name", "math_score"]]` |
| `df.loc[row, col]` | Select by **label** (name/index) | `df.loc[0, "name"]` |
| `df.iloc[row, col]` | Select by **position number** | `df.iloc[0, 0]` |
| `df[condition]` | Filter rows by condition | `df[df["math_score"] > 70]` |

> 🔑 **loc vs iloc:**
> - `loc` uses the **label** — what the row/column is called
> - `iloc` uses the **position** — the number 0, 1, 2... from the top-left

````python
# Example 1: Selecting columns

import pandas as pd

df = pd.read_csv("data/students.csv")

# Select a single column — returns a Series
print("Just the names:")
print(df["name"])
print()

# Select multiple columns — returns a DataFrame
print("Names and math scores:")
print(df[["name", "math_score", "english_score"]])
````

````python
# Example 2: Filtering rows with conditions
# Scenario: Find students who scored above 75 in maths AND have paid their fees.

import pandas as pd

df = pd.read_csv("data/students.csv")

# Single condition
high_scorers = df[df["math_score"] > 75]
print("Students scoring above 75 in maths:")
print(high_scorers[["name", "city", "math_score"]])
print()

# Multiple conditions — use & (and) / | (or) with brackets around each condition
top_paid = df[(df["math_score"] > 75) & (df["fees_paid"] == True)]
print("High scorers who have paid fees:")
print(top_paid[["name", "math_score", "fees_paid"]])
````

````python
# Example 3: loc and iloc — precision selection

import pandas as pd

df = pd.read_csv("data/students.csv")

# loc — select by label (row index label, column name)
print("Row 0, name column (loc):")
print(df.loc[0, "name"])           # first row, name column

print("\nRows 0-2, name and city (loc):")
print(df.loc[0:2, ["name", "city"]])  # rows 0 to 2 inclusive

# iloc — select by position number
print("\nFirst row, first column (iloc):")
print(df.iloc[0, 0])               # row index 0, column index 0

print("\nLast 2 rows, first 3 columns (iloc):")
print(df.iloc[-2:, :3])
````

## 6. Data Manipulation and Transformation

### Concept Explanation

Data manipulation means **changing, adding, or transforming** data to make it more useful for analysis. Think of it like cooking — you take raw ingredients (raw data) and prepare them into something more useful.

**Common transformation operations:**

| Operation | Code | What it does |
|-----------|------|--------------|
| Add new column | `df["new_col"] = expression` | Create a column from existing ones |
| Rename columns | `df.rename(columns={"old": "new"})` | Rename column headers |
| Apply a function | `df["col"].apply(func)` | Apply a function to every value in a column |
| Replace values | `df["col"].replace({old: new})` | Swap specific values |
| Convert types | `df["col"].astype(int)` | Change a column's data type |

````python
# Example 1: Create new columns from existing ones
# Scenario: Calculate each student's average score across 3 subjects.

import pandas as pd

df = pd.read_csv("data/students.csv")

# Fill missing science scores first
df["science_score"] = df["science_score"].fillna(df["science_score"].mean())

# Create a new column: average of the three subject scores
df["avg_score"] = (df["math_score"] + df["english_score"] + df["science_score"]) / 3
df["avg_score"] = df["avg_score"].round(1)   # round to 1 decimal place

# Create a pass/fail column: pass if avg_score >= 50
df["status"] = df["avg_score"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print(df[["name", "math_score", "english_score", "science_score", "avg_score", "status"]])
````

````python
# Example 2: apply() with a custom function — assign letter grades

import pandas as pd

df = pd.read_csv("data/students.csv")
df["science_score"] = df["science_score"].fillna(df["science_score"].mean())
df["avg_score"] = ((df["math_score"] + df["english_score"] + df["science_score"]) / 3).round(1)

# Define a function that converts a score to a letter grade
def assign_grade(score):
    if score >= 80:   return "A"
    elif score >= 60: return "B"
    elif score >= 40: return "C"
    else:             return "D"

# Apply that function to every value in avg_score
df["grade"] = df["avg_score"].apply(assign_grade)

print(df[["name", "avg_score", "grade"]])
````

````python
# Example 3: Rename columns and convert types

import pandas as pd

df = pd.read_csv("data/students.csv")

# Rename columns to be cleaner
df = df.rename(columns={
    "math_score"    : "Maths",
    "english_score" : "English",
    "science_score" : "Science",
    "attendance_pct": "Attendance (%)"
})

print("Renamed columns:", list(df.columns))
print()

# fees_paid is boolean — convert it to a readable string
df["fees_paid"] = df["fees_paid"].replace({True: "Paid", False: "Owing"})

print(df[["name", "fees_paid"]].head())
````

## 7. Data Summarisation and Aggregation

### Concept Explanation

**Aggregation** means collapsing many rows of data into summary numbers — like calculating the average score per city, or the total students per school.

> 🎯 **Analogy:** Instead of reading every student's score one by one, you ask: *"What is the average score per school?"* — that's aggregation.

The most powerful aggregation tool is **`groupby()`** — it splits the data into groups, applies a calculation to each group, and combines the results.

```python
df.groupby("city")["math_score"].mean()  # average math score per city
```

**Common aggregation functions:**
`mean()`, `sum()`, `count()`, `min()`, `max()`, `std()`, `agg({"col": ["mean", "max"]})`

````python
# Example 1: groupby() with a single aggregation
# Scenario: What is the average maths score per city?

import pandas as pd

df = pd.read_csv("data/students.csv")

# Group by city, calculate mean math_score for each city
avg_by_city = df.groupby("city")["math_score"].mean().round(1)

print("Average maths score by city:")
print(avg_by_city)
````

````python
# Example 2: Multiple aggregations with agg()
# Scenario: For each school, get the count, mean, min, and max maths score.

import pandas as pd

df = pd.read_csv("data/students.csv")

school_summary = df.groupby("school")["math_score"].agg(
    total_students = "count",
    average_score  = "mean",
    lowest_score   = "min",
    highest_score  = "max"
).round(1)

print("Maths performance summary by school:")
print(school_summary)
````

````python
# Example 3: Groupby with multiple columns
# Scenario: How many students per city have paid vs not paid fees?

import pandas as pd

df = pd.read_csv("data/students.csv")

fees_by_city = df.groupby(["city", "fees_paid"])["name"].count()
fees_by_city = fees_by_city.rename("student_count")

print("Fee payment status by city:")
print(fees_by_city)
````

## 8. Sorting and Reshaping Data

### Concept Explanation

**Sorting** arranges rows in a specific order — like ranking students from highest to lowest score. **Reshaping** changes the *structure* of the table — like turning a list of scores into a grid where each row is a city and each column is a subject.

| Function | What it does |
|----------|--------------|
| `df.sort_values("col")` | Sort by a column (ascending by default) |
| `df.sort_values("col", ascending=False)` | Sort descending (highest first) |
| `df.sort_values(["col1", "col2"])` | Sort by multiple columns |
| `df.reset_index(drop=True)` | Reset row numbers after sorting/filtering |
| `df.pivot_table(...)` | Create a summary grid (like an Excel pivot table) |

````python
# Example 1: Sorting — rank students by maths score

import pandas as pd

df = pd.read_csv("data/students.csv")

# Sort by math_score — highest first
ranked = df.sort_values("math_score", ascending=False).reset_index(drop=True)

# Add a rank column
ranked["rank"] = ranked.index + 1   # index starts at 0, so add 1

print("Maths leaderboard:")
print(ranked[["rank", "name", "school", "math_score"]])
````

````python
# Example 2: Sorting by multiple columns
# Scenario: Sort by city (A-Z), then by maths score (highest first) within each city.

import pandas as pd

df = pd.read_csv("data/students.csv")

sorted_df = df.sort_values(
    by=["city", "math_score"],
    ascending=[True, False]    # city A-Z, score high-to-low
).reset_index(drop=True)

print(sorted_df[["name", "city", "math_score"]])
````

````python
# Example 3: pivot_table() — reshape data into a summary grid
# Scenario: What is the average score per subject per city?

import pandas as pd

df = pd.read_csv("data/students.csv")
df["science_score"] = df["science_score"].fillna(df["science_score"].mean())

# pivot_table creates a cross-tab summary
pivot = df.pivot_table(
    index   = "city",                                         # rows
    values  = ["math_score", "english_score", "science_score"],  # columns
    aggfunc = "mean"                                          # aggregation
).round(1)

print("Average scores per city (pivot table):")
print(pivot)
````

## 9. Combining Datasets

### Concept Explanation

In real projects, data often comes in **separate tables** that you need to combine. Pandas has two main ways:

---

### `pd.concat()` — Stack tables on top of each other
Use when both tables have the **same columns** and you want to join them into one longer table.
```
Table A        Table B        Result
Achieng   +   Wanjiru   =   Achieng
Kamau         Ochieng       Kamau
                            Wanjiru
                            Ochieng
```

### `pd.merge()` — Join tables side by side using a shared column
Use when two tables share a **common column** (like a student ID or name) and you want to combine their information.
```
Students table     +   Scores table   =   Combined
name | city            name | score       name | city | score
Achieng | Nairobi      Achieng | 85        Achieng | Nairobi | 85
```

> 🎯 **Analogy:**
> - `concat` = stacking exercise books from two classes into one pile
> - `merge` = matching a student's name on the fee receipt with the same name on the results sheet

````python
# Example 1: pd.concat() — combine two class lists
# Scenario: Two bootcamp cohorts are merged into one master list.

import pandas as pd

cohort_a = pd.DataFrame({
    "name"  : ["Achieng", "Kamau", "Zawadi"],
    "city"  : ["Nairobi", "Mombasa", "Kisumu"],
    "cohort": ["A", "A", "A"]
})

cohort_b = pd.DataFrame({
    "name"  : ["Wanjiru", "Ochieng", "Baraka"],
    "city"  : ["Eldoret", "Nakuru", "Kisumu"],
    "cohort": ["B", "B", "B"]
})

# Stack vertically — ignore_index resets the row numbers
all_students = pd.concat([cohort_a, cohort_b], ignore_index=True)

print("Combined master list:")
print(all_students)
````

````python
# Example 2: pd.merge() — join students with their scores
# Scenario: The admin team has student details in one table and exam scores in another.

import pandas as pd

# Table 1: student details
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105],
    "name"      : ["Achieng", "Kamau", "Zawadi", "Odhiambo", "Njeri"],
    "city"      : ["Nairobi", "Mombasa", "Kisumu", "Nakuru", "Eldoret"]
})

# Table 2: exam scores (note: Odhiambo is missing — he was absent)
scores = pd.DataFrame({
    "student_id" : [101, 102, 103, 105],
    "math_score" : [85, 72, 91, 63],
    "english_score": [78, 65, 85, 71]
})

# Inner merge — only students who appear in BOTH tables
merged = pd.merge(students, scores, on="student_id", how="inner")

print("Inner merge (only students with scores):")
print(merged)
````

````python
# Example 3: Left merge — keep ALL students, fill missing scores with NaN

import pandas as pd

students = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105],
    "name"      : ["Achieng", "Kamau", "Zawadi", "Odhiambo", "Njeri"],
    "city"      : ["Nairobi", "Mombasa", "Kisumu", "Nakuru", "Eldoret"]
})

scores = pd.DataFrame({
    "student_id"   : [101, 102, 103, 105],
    "math_score"   : [85, 72, 91, 63],
    "english_score": [78, 65, 85, 71]
})

# Left merge — keep ALL students from 'students' table
# Odhiambo will appear with NaN scores since he has no entry in scores table
left_merged = pd.merge(students, scores, on="student_id", how="left")

print("Left merge (all students, NaN where no score):")
print(left_merged)
````

## 10. Data Visualisation with Pandas

### Concept Explanation

Numbers in a table are useful, but a **chart** makes patterns immediately visible. Pandas has a built-in `.plot()` method that creates charts directly from a DataFrame.

**Common chart types:**

| Chart type | Code | Best for |
|------------|------|---------|
| Bar chart | `df.plot(kind="bar")` | Comparing categories (e.g., scores per school) |
| Horizontal bar | `df.plot(kind="barh")` | Many categories — easier to read labels |
| Line chart | `df.plot(kind="line")` | Trends over time |
| Histogram | `df["col"].plot(kind="hist")` | Distribution of one column |
| Scatter plot | `df.plot(kind="scatter", x="col1", y="col2")` | Relationship between two columns |
| Pie chart | `df.plot(kind="pie")` | Parts of a whole |

> 💡 Always add a **title** and **axis labels** to make your chart understandable to others!

````python
# Example 1: Bar chart — average maths score per city

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/students.csv")

# Calculate average maths score per city
avg_by_city = df.groupby("city")["math_score"].mean().sort_values(ascending=False)

# Plot as a horizontal bar chart
avg_by_city.plot(kind="barh", color="steelblue", edgecolor="black")

plt.title("Average Maths Score by City")
plt.xlabel("Average Score")
plt.ylabel("City")
plt.tight_layout()   # prevents labels from being cut off
plt.show()
````

````python
# Example 2: Histogram — distribution of maths scores
# A histogram shows how scores are spread — are most students scoring in the middle?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/students.csv")

df["math_score"].plot(
    kind="hist",
    bins=5,           # divide scores into 5 buckets
    color="orange",
    edgecolor="black"
)

plt.title("Distribution of Maths Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
````

````python
# Example 3: Scatter plot — does attendance affect scores?
# A scatter plot shows whether two things move together.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/students.csv")

df.plot(
    kind="scatter",
    x="attendance_pct",
    y="math_score",
    color="green",
    alpha=0.7         # slight transparency so overlapping dots are visible
)

plt.title("Attendance vs Maths Score")
plt.xlabel("Attendance (%)")
plt.ylabel("Maths Score")
plt.tight_layout()
plt.show()

# Does a higher attendance tend to go with a higher score?
correlation = df["attendance_pct"].corr(df["math_score"])
print(f"Correlation between attendance and maths score: {correlation:.2f}")
print("(A value near +1 means strong positive relationship)")
````
