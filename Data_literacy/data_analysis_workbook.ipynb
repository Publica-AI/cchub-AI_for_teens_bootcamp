Read through this carefully **before** you run the cell. Each line has one specific job:

grade_counts.plot(

## Step 8 — Does Studying More = Better Grades? 📚

This is one of the most common questions in education research.

**Our approach:**  
We will calculate the **average study time** for students in each grade group and compare them.

We use `.groupby()` — this groups all students by their grade, then calculates something for each group separately.

Think of it like separating students into 5 rooms (one per grade), then asking *"what is the average study time in each room?"*

```python
df.groupby('GradeClass')['StudyTimeWeekly'].mean()
#          ↑ group by     ↑ column to look at  ↑ calculate average
```

````python
# Average study time for each grade group
study_by_grade = df.groupby('GradeClass')['StudyTimeWeekly'].mean().round(2)

# Add readable grade labels
study_by_grade.index = study_by_grade.index.map({0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'F'})

print('Average weekly study hours by grade:')
print(study_by_grade)
````

#### 📊 Now Let's Chart It

We already imported `matplotlib` in Step 7 — so `plt` is ready to go, no need to import again.

This chart is built the same way as before. The only thing changing is the **data** (`study_by_grade` instead of `grade_counts`) and the **labels** to match what we are showing.

Notice `fontsize=14` inside `plt.title()` — that just controls how big the title text is. Try changing it to `20` or `10` and see what happens!

````python
study_by_grade.plot(
    kind='bar',
    color=['#2ecc71', '#3498db', '#f39c12', '#e67e22', '#e74c3c'],
    edgecolor='black'
)

plt.title('Average Study Time by Grade', fontsize=14)
plt.xlabel('Grade')
plt.ylabel('Average Hours per Week')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

## Step 10 — Does Having a Tutor Help? 📝

Some students have private tutors (`Tutoring = 1`) and some don't (`Tutoring = 0`).

Let's compare the **average GPA** of tutored vs non-tutored students.

Same idea — `groupby` on `Tutoring`, calculate `mean` of `GPA`.

````python
# Compare GPA between tutored and non-tutored students
gpa_by_tutoring = df.groupby('Tutoring')['GPA'].mean().round(3)

# Add readable labels
gpa_by_tutoring.index = gpa_by_tutoring.index.map({0: 'No Tutor', 1: 'Has Tutor'})

print('Average GPA:')
print(gpa_by_tutoring)

difference = gpa_by_tutoring['Has Tutor'] - gpa_by_tutoring['No Tutor']
print(f'\nDifference: {difference:.3f} GPA points')
````

#### 📊 Two Bars, One New Setting

This time we are only comparing **two groups** — tutor vs no tutor. With just two bars, the chart can look a bit wide and stretched. That is why we added one new setting:

```python
width=0.4    ← controls how WIDE each bar is (default is 0.8)
             ← making them narrower looks cleaner when there are only 2 bars
```

We also added:

```python
plt.ylim(0, 4)    ← sets the y-axis range from 0 to 4 (the full GPA scale)
                  ← without this, matplotlib might zoom in and make a tiny
                     difference look huge — which would be misleading!
```

> 💡 **Always check your y-axis scale.** A chart that starts at 2.0 instead of 0 can make a small difference look dramatic. Good data analysts always keep the full scale in mind.

````python
gpa_by_tutoring.plot(
    kind='bar',
    color=['#3498db', '#e74c3c'],
    edgecolor='black',
    width=0.4
)

plt.title('Average GPA: Tutored vs Not Tutored', fontsize=14)
plt.ylabel('Average GPA')
plt.xticks(rotation=0)
plt.ylim(0, 4)
plt.tight_layout()
plt.show()
````

## Step 11 — Sports, Music, or Volunteering — Who Gets Better Grades? 🎯

There is a big debate: should students focus only on academics, or do activities like sports and music actually help?

Let's check the data for all three activities at once.

Here we use a **loop**(I believe you still remember this from python class) — instead of writing the same code three times, we loop through each activity and print the result.

````python
activities = ['Sports', 'Music', 'Volunteering']

print('Average GPA by activity participation:')
print('-' * 45)

for activity in activities:
    gpa_no  = df[df[activity] == 0]['GPA'].mean()
    gpa_yes = df[df[activity] == 1]['GPA'].mean()
    diff    = gpa_yes - gpa_no
    direction = 'higher' if diff > 0 else 'lower'
    print(f'{activity}:')
    print(f'  No  → avg GPA: {gpa_no:.3f}')
    print(f'  Yes → avg GPA: {gpa_yes:.3f}')
    print(f'  Participants score {abs(diff):.3f} points {direction}')
    print()
````

## Step 12 — Does Parental Support Matter? 👨‍👩‍👦

The `ParentalSupport` column measures how much support students get at home:

- **0** = None
- **1** = Low
- **2** = Moderate  
- **3** = High
- **4** = Very High

Let's see how GPA changes as parental support increases.

````python
support_labels = {0: 'None', 1: 'Low', 2: 'Moderate', 3: 'High', 4: 'Very High'}

gpa_by_support = df.groupby('ParentalSupport')['GPA'].mean().round(3)
gpa_by_support.index = gpa_by_support.index.map(support_labels)

print('Average GPA by parental support level:')
print(gpa_by_support)
````

#### 📊 A New Chart Type — The Line Chart

For this one we are switching from a **bar chart** to a **line chart**. Here is why:

- A **bar chart** is great when your groups are separate categories (like A, B, C, D, F — they don't flow into each other).
- A **line chart** is better when your categories have a natural **order** and you want to see a **trend** — like going from None → Low → Moderate → High → Very High. The line connecting the dots shows you whether the value is going up or down as you move along.

Here are the new settings you will see for the first time:

```python
kind='line'       ← draw a line chart instead of bars
marker='o'        ← put a small circle (dot) at each data point so it is easy to see
linewidth=2.5     ← how thick the line is (try 1 for thin, 5 for thick)
markersize=8      ← how big the dots are at each data point
```

And one brand new `plt` command:

```python
plt.grid(axis='y', linestyle='--', alpha=0.5)
#         ↑ only draw grid lines going ACROSS (horizontal)
#                    ↑ make them dashed lines
#                                  ↑ alpha controls transparency: 0 = invisible, 1 = solid
#                                    0.5 = light/faint so it doesn't distract
```

Grid lines are helpful on a line chart because they make it much easier to read off the exact GPA value for each point.

````python
gpa_by_support.plot(
    kind='line',
    marker='o',
    color='#8e44ad',
    linewidth=2.5,
    markersize=8
)

plt.title('Does Parental Support Affect GPA?', fontsize=14)
plt.xlabel('Level of Parental Support')
plt.ylabel('Average GPA')
plt.ylim(0, 4)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## Step 13 — GPA Distribution: What Does It Really Look Like? 📈

So far we have been looking at **averages**. But averages can hide a lot.

For example: if one student scores 4.0 and another scores 0.0, the average is 2.0 — but that doesn't mean everyone is a C student. The **distribution** shows you the full picture.

A **histogram** is the chart for this. Think of it like lining up all 2,392 students side by side based on their GPA — the taller the bar, the more students scored in that GPA range.

#### 📊 A Third Chart Type — The Histogram

The histogram uses `kind='hist'`. Let's look at its two new settings:

```python
bins=30    ← split the GPA range (0 to 4) into 30 equal-sized buckets
           ← each bar represents one bucket
           ← more bins = more detail but more jagged
           ← fewer bins = smoother but less detail
           ← try changing 30 to 10 or 50 and see how the shape changes!
```

And there is one completely new `plt` command here — a **reference line**:

```python
plt.axvline(
    df['GPA'].mean(),     ← draw a vertical line at the average GPA value
    color='red',          ← make it red so it stands out
    linestyle='--',       ← dashed line
    label=f"Average: {df['GPA'].mean():.2f}"   ← the text shown in the legend
)
plt.legend()   ← show the legend box (so the label appears on the chart)
```

`axvline` stands for **axis vertical line** — it draws a vertical line across the chart at any x-value you choose. We are using it to mark the average GPA so you can instantly see where it falls compared to the rest of the students.

````python
df['GPA'].plot(
    kind='hist',
    bins=30,
    color='#3498db',
    edgecolor='white'
)

plt.title('Distribution of Student GPAs', fontsize=14)
plt.xlabel('GPA')
plt.ylabel('Number of Students')
plt.axvline(df['GPA'].mean(), color='red', linestyle='--', label=f"Average: {df['GPA'].mean():.2f}")
plt.legend()
plt.tight_layout()
plt.show()
````
