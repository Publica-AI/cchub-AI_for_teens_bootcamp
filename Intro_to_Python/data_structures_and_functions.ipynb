## 1. Introduction to Data Structures

### Concept Explanation

So far, each variable has held **one value** at a time:
```python
student1 = "Achieng"
student2 = "Kamau"
student3 = "Zawadi"
```

Imagine doing this for a class of 40 students — that's 40 separate variables! A **data structure** solves this by letting one variable hold **many values** at once.

Python has three main data structures you will learn:

| Data Structure | Description | Example |
|----------------|-------------|------------------------------------------------------|
| **List** | Ordered, changeable collection | `["Achieng", "Kamau", "Zawadi"]` |
| **Tuple** | Ordered, **unchangeable** collection | `("Nairobi", -1.286, 36.817)` |
| **Dictionary** | Key–value pairs (like a contacts list) | `{"name": "Mwangi", "age": 16}` |

> 🗂️ **Quick way to remember:**
> - **List** → a shopping list you can edit
> - **Tuple** → your date of birth — it never changes
> - **Dictionary** → your phone contacts — each name (key) links to a number (value)

````python
# A quick preview of all three data structures

# List — a changeable collection of student names
students = ["Achieng", "Kamau", "Zawadi", "Odhiambo"]

# Tuple — fixed coordinates of Nairobi city centre (latitude, longitude)
nairobi_coords = (-1.286, 36.817)

# Dictionary — a student's profile with named fields
student_profile = {"name": "Mwangi", "age": 16, "city": "Nakuru"}

print("Students list  :", students)
print("Nairobi coords :", nairobi_coords)
print("Student profile:", student_profile)
````

## 2. Tuples

### Concept Explanation

A **tuple** is an ordered collection of values that **cannot be changed** once created. It is defined using **round brackets `()`**.

```python
my_tuple = (value1, value2, value3)
```

**Why use a tuple instead of a list?**
Use a tuple when the data should *stay fixed* — like the days of the week, GPS coordinates, or a student's date of birth. Trying to change a tuple will cause an error, which protects your data from accidental changes.

**Accessing items — indexing:**
Items in a tuple are numbered starting from **0** (not 1).
```
tuple:  ("Mon", "Tue", "Wed")
index:    [0]    [1]    [2]
```

> 💡 A tuple with **one item** needs a trailing comma: `("Mombasa",)` — without it, Python thinks it's just a value in brackets.

````python
# Example 1: Creating and accessing a tuple
# Scenario: Store the top 3 cities by population — this ranking rarely changes.

top_cities = ("Nairobi", "Mombasa", "Kisumu")

# Access items using their index (position), starting from 0
print("1st city:", top_cities[0])   # Nairobi
print("2nd city:", top_cities[1])   # Mombasa
print("3rd city:", top_cities[2])   # Kisumu

# len() tells you how many items are in the tuple
print("Number of cities:", len(top_cities))
````

````python
# Example 2: Tuple unpacking
# Scenario: Store Zawadi's date of birth as a tuple and unpack it into variables.

zawadi_dob = (15, 8, 2009)   # day, month, year — fixed forever!

# Tuple unpacking assigns each value to a separate variable in one line
day, month, year = zawadi_dob

print(f"Zawadi was born on day {day}, month {month}, year {year}.")

# You can also loop through a tuple
rgb_orange = (255, 165, 0)   # RGB colour values never change for a specific colour
labels = ("Red", "Green", "Blue")

for label, value in zip(labels, rgb_orange):
    print(f"  {label}: {value}")
````

````python
# Example 3: Tuples are immutable — trying to change one causes an error
# Scenario: Demonstrate what happens when you try to change a tuple.

school_terms = ("Term 1", "Term 2", "Term 3")   # school calendar is fixed

print("School terms:", school_terms)
print("First term  :", school_terms[0])

# Uncomment the line below to see the error — tuples cannot be changed!
# school_terms[0] = "Vacation"   # TypeError: 'tuple' object does not support item assignment
````

## 3. Lists

### Concept Explanation

A **list** is an ordered collection of values that **can be changed** — you can add, remove, or update items. Lists use **square brackets `[]`**.

```python
my_list = [value1, value2, value3]
```

Lists support **indexing** just like tuples — items are numbered from `0`. They also support **negative indexing**: `-1` is the last item, `-2` is second to last, etc.

**Common list methods:**

| Method | What it does | Example |
|--------|--------------|---------|
| `.append(x)` | Adds `x` to the end | `names.append("Njeri")` |
| `.remove(x)` | Removes the first occurrence of `x` | `names.remove("Kamau")` |
| `.sort()` | Sorts the list in place | `scores.sort()` |
| `.reverse()` | Reverses the list | `scores.reverse()` |
| `.pop()` | Removes and returns the last item | `names.pop()` |
| `len(list)` | Returns the number of items | `len(names)` |

````python
# Example 1: Creating and accessing a list
# Scenario: A class register of student names.

class_register = ["Achieng", "Baraka", "Chebet", "Fatuma", "Kamau"]

# Access by index (starts at 0)
print("First student :", class_register[0])    # Achieng
print("Last student  :", class_register[-1])   # Kamau (negative index)

# Loop through the entire list
print("\nFull register:")
for i, name in enumerate(class_register, start=1):   # enumerate adds a counter
    print(f"  {i}. {name}")
````

````python
# Example 2: Modifying a list — add, update, remove
# Scenario: The canteen menu changes throughout the week.

menu = ["ugali", "chapati", "mandazi"]
print("Original menu:", menu)

# Add a new item to the end
menu.append("pilau")
print("After append  :", menu)

# Update an item by reassigning its index
menu[1] = "nyama choma"        # replace chapati with nyama choma
print("After update  :", menu)

# Remove a specific item
menu.remove("mandazi")
print("After remove  :", menu)
````

````python
# Example 3: Useful list operations — sort, sum, min, max
# Scenario: Analyse exam scores for a group of students.

scores = [72, 45, 88, 61, 95, 53, 77]

print("Original scores:", scores)

scores.sort()   # sort from lowest to highest
print("Sorted scores  :", scores)

# Built-in functions work on lists
print("Highest score  :", max(scores))
print("Lowest score   :", min(scores))
print("Total marks    :", sum(scores))
print("Average score  :", sum(scores) / len(scores))
````

## 4. Dictionaries

### Concept Explanation

A **dictionary** stores data as **key–value pairs** — like a real dictionary where a word (key) has a definition (value), or a contacts list where a name (key) links to a phone number (value).

Dictionaries use **curly brackets `{}`**:
```python
my_dict = {"key1": value1, "key2": value2}
```

- You access values using their **key**, not an index number.
- Keys must be **unique** — no two entries can have the same key.
- Values can be any data type: strings, numbers, lists, even other dictionaries.

**Common dictionary methods:**

| Method / Operation | What it does |
|--------------------|--------------|
| `dict["key"]` | Get the value for a key |
| `dict["key"] = value` | Add or update a key–value pair |
| `del dict["key"]` | Delete a key–value pair |
| `.keys()` | Returns all keys |
| `.values()` | Returns all values |
| `.items()` | Returns all key–value pairs |

````python
# Example 1: Creating and accessing a dictionary
# Scenario: Store a student's profile with named fields.

student = {
    "name"    : "Wanjiru",
    "age"     : 16,
    "city"    : "Nyeri",
    "score"   : 84,
    "passed"  : True
}

# Access values using their key (not a number!)
print("Name  :", student["name"])
print("Age   :", student["age"])
print("Passed:", student["passed"])

# Print a full sentence using the dictionary
print(f"\n{student['name']} from {student['city']} scored {student['score']}.")
````

````python
# Example 2: Adding, updating, and deleting entries
# Scenario: Manage a canteen price list that changes sometimes.

canteen_prices = {
    "chapati"  : 15,
    "mandazi"  : 10,
    "pilau"    : 80,
    "juice"    : 50
}

print("Original prices:", canteen_prices)

# Add a new item
canteen_prices["chai"] = 20
print("After adding chai  :", canteen_prices)

# Update an existing price
canteen_prices["chapati"] = 20   # price went up
print("After updating chapati:", canteen_prices)

# Remove an item that is no longer sold
del canteen_prices["mandazi"]
print("After removing mandazi:", canteen_prices)
````

````python
# Example 3: Looping through a dictionary
# Scenario: Print a formatted menu using .items()

canteen_prices = {
    "Ugali and sukuma" : 60,
    "Chapati"          : 20,
    "Pilau"            : 80,
    "Chai"             : 20,
    "Juice"            : 50
}

print("======= TODAY'S MENU =======")
for item, price in canteen_prices.items():   # .items() gives both key and value
    print(f"  {item:<22} Ksh {price}")
print("============================")
````

## 5. Introduction to Functions

### Concept Explanation

A **function** is a named, reusable block of code that performs a specific task. Instead of writing the same code over and over, you write it once inside a function and *call* it by name whenever you need it.

> 🍳 **Analogy:** A recipe for chapatis is a function. You write the steps once. Whenever you want chapatis, you follow the recipe — you don't rewrite it each time!

**Defining a function:**
```python
def function_name():
    # code that runs when you call this function
```

**Calling a function:**
```python
function_name()   # this runs the code inside
```

> 🔑 **Key rules:**
> - Start with the keyword `def` (short for *define*).
> - The function name is followed by round brackets `()` and a colon `:`.
> - The code inside must be indented.
> - **Defining** a function does NOT run it — you must **call** it.

````python
# Example 1: Defining and calling a basic function
# Scenario: A greeting function for new bootcamp students.

def greet_student():
    # This code runs every time greet_student() is called
    print("Welcome to the coding bootcamp!")
    print("Let's build something amazing today.")

# Call the function — you can call it as many times as you like
greet_student()
print("---")
greet_student()
````

````python
# Example 2: A function that performs a calculation
# Scenario: Calculate the total bill at the canteen.

def calculate_canteen_bill():
    chapati  = 20    # price per chapati
    juice    = 50    # price of juice
    quantity = 2     # number of chapatis

    total = (chapati * quantity) + juice
    print(f"Total canteen bill: Ksh {total}")

# Call the function
calculate_canteen_bill()
````

````python
# Example 3: Why functions are useful — avoiding repetition
# Scenario: Print a divider line before and after any section header.

def print_divider():
    print("=" * 35)   # prints 35 equal signs

# Without the function you'd write print("=" * 35) every single time
print_divider()
print("     STUDENT RESULTS — TERM 2     ")
print_divider()
print("Kamau    : 78")
print("Wanjiru  : 91")
print("Odhiambo : 64")
print_divider()
````

## 6. Function Parameters

### Concept Explanation

**Parameters** make functions flexible by letting you pass information *into* the function when you call it. Without parameters, a function always does exactly the same thing. With parameters, it can work with different data each time.

> 🎯 **Analogy:** The chapati recipe is the function. The *number of chapatis* you want is the parameter — you pass it in and the recipe adjusts!

```python
def function_name(parameter1, parameter2):
    # use parameter1 and parameter2 inside
```

When you **call** the function, the values you pass in are called **arguments**:
```python
function_name(value1, value2)   # value1 → parameter1, value2 → parameter2
```

**Default parameters** give a parameter a fallback value if none is provided:
```python
def greet(name, greeting="Hello"):
    print(greeting, name)
```

````python
# Example 1: Function with one parameter
# Scenario: Greet any student by name.

def greet_student(name):           # 'name' is the parameter
    print(f"Hello, {name}! Welcome to class.")

# Call with different arguments each time
greet_student("Achieng")          # 'Achieng' is the argument
greet_student("Baraka")
greet_student("Fatuma")
````

````python
# Example 2: Function with multiple parameters
# Scenario: Calculate the matatu fare for any route and number of passengers.

def matatu_fare(route, passengers, fare_per_person):
    total = passengers * fare_per_person
    print(f"Route: {route}")
    print(f"  {passengers} passenger(s) × Ksh {fare_per_person} = Ksh {total}")

# Call with different routes and passenger counts
matatu_fare("Nairobi to Thika", 3, 80)
print()
matatu_fare("Kisumu to Eldoret", 5, 250)
````

````python
# Example 3: Default parameter values
# Scenario: Assign a grade with a default pass mark of 50.

def check_result(student_name, score, pass_mark=50):   # pass_mark defaults to 50
    if score >= pass_mark:
        status = "PASS"
    else:
        status = "FAIL"
    print(f"{student_name}: {score} — {status}")

# Uses default pass mark of 50
check_result("Zawadi", 73)
check_result("Mwangi", 41)

# Override the default pass mark with 60
check_result("Njeri", 58, 60)    # now 60 is needed to pass
````

## 7. Return Values

### Concept Explanation

So far, our functions have only **printed** results. But printing is not always enough — sometimes you want to **use the result** in another part of your program.

The `return` keyword sends a value **back to the caller**, where it can be stored in a variable or used in an expression.

> 🎯 **Analogy:** You send a message to a friend asking the price of unga. They `return` the answer to you — now you have the information to use however you like. If they only shouted the price to themselves, it would be useless to you!

```python
def add(a, b):
    return a + b      # sends the result back

result = add(3, 4)   # result now holds 7
print(result)        # 7
```

> ⚠️ Once Python hits `return`, the function **stops immediately**. Any code after `return` in the same function will not run.

````python
# Example 1: A function that returns a calculated value
# Scenario: Calculate and RETURN the total canteen bill (don't just print it).

def calculate_total(price, quantity):
    total = price * quantity
    return total    # sends the total back to whoever called this function

# Now we can USE the returned value
chapati_bill = calculate_total(20, 3)    # Ksh 20 × 3 chapatis
juice_bill   = calculate_total(50, 2)    # Ksh 50 × 2 juices

grand_total = chapati_bill + juice_bill  # we can do maths with returned values!

print(f"Chapati bill : Ksh {chapati_bill}")
print(f"Juice bill   : Ksh {juice_bill}")
print(f"Grand total  : Ksh {grand_total}")
````

````python
# Example 2: A function that returns a grade letter
# Scenario: Return a grade so the result can be used elsewhere in the program.

def get_grade(score):
    if score >= 80:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 40:
        return "C"
    else:
        return "D"

# Use the returned grade in a sentence
amina_score   = 76
amina_grade   = get_grade(amina_score)   # returns "B"

baraka_score  = 91
baraka_grade  = get_grade(baraka_score)  # returns "A"

print(f"Amina  scored {amina_score}  → Grade {amina_grade}")
print(f"Baraka scored {baraka_score} → Grade {baraka_grade}")
````

````python
# Example 3: Returning multiple values at once (as a tuple)
# Scenario: Calculate both the average score AND the highest score for a class.

def analyse_scores(scores):
    average = sum(scores) / len(scores)
    highest = max(scores)
    lowest  = min(scores)
    return average, highest, lowest    # returns a tuple of 3 values

class_scores = [72, 85, 60, 91, 55, 78, 88]

# Unpack the three returned values
avg, high, low = analyse_scores(class_scores)

print(f"Average score : {avg:.1f}")
print(f"Highest score : {high}")
print(f"Lowest score  : {low}")
````

## 8. Using Functions with Data Structures

### Concept Explanation

Functions and data structures become truly powerful when you **combine** them. You can pass a list or dictionary *into* a function, process it, and return a result. This lets you write clean, reusable code for real-world problems.

> 🎯 **Think of it this way:** The data structure is the *ingredients* (a list of students, a dictionary of prices). The function is the *chef* — it takes the ingredients, does something useful with them, and serves up a result.

Common patterns:
- Pass a **list** to a function → process every item
- Pass a **dictionary** to a function → read or update entries
- **Return** a new list or dictionary from a function

````python
# Example 1: Pass a list to a function and process it
# Scenario: Print a class report with grades for every student in a list.

def get_grade(score):
    if score >= 80: return "A"
    elif score >= 60: return "B"
    elif score >= 40: return "C"
    else: return "D"

def print_class_report(results):
    print("====== CLASS REPORT ======")
    for name, score in results:
        grade = get_grade(score)    # call get_grade inside the loop!
        print(f"  {name:<12} {score}   Grade {grade}")
    print("==========================")

# A list of (name, score) tuples
class_results = [
    ("Achieng",  88),
    ("Kamau",    55),
    ("Zawadi",   72),
    ("Baraka",   38),
    ("Wanjiru",  91)
]

print_class_report(class_results)
````

````python
# Example 2: Pass a dictionary to a function
# Scenario: Calculate the total bill from a canteen order dictionary.

def calculate_order_total(order, prices):
    """
    order  : dict of {item: quantity}
    prices : dict of {item: price}
    Returns the grand total cost.
    """
    total = 0
    for item, quantity in order.items():
        cost   = prices[item] * quantity
        total += cost
        print(f"  {item:<15} × {quantity}  = Ksh {cost}")
    return total

# Price list
canteen_prices = {"chapati": 20, "juice": 50, "mandazi": 10, "pilau": 80}

# Odhiambo's order
odhiambo_order = {"chapati": 2, "juice": 1, "mandazi": 3}

print("Odhiambo's order:")
bill = calculate_order_total(odhiambo_order, canteen_prices)
print(f"  {'TOTAL':<15}         Ksh {bill}")
````

````python
# Example 3: Function that RETURNS a new list (filtered results)
# Scenario: From a list of students, return only those who passed.

def get_passing_students(results, pass_mark=50):
    passing = []   # start with an empty list
    for name, score in results:
        if score >= pass_mark:
            passing.append(name)   # add to list only if they passed
    return passing   # return the filtered list

all_results = [
    ("Chebet",   82),
    ("Otieno",   44),
    ("Fatuma",   67),
    ("Muthoni",  31),
    ("Ochieng",  75)
]

passing_students = get_passing_students(all_results)

print("Students who passed:")
for student in passing_students:
    print(f"  ✓ {student}")
````
