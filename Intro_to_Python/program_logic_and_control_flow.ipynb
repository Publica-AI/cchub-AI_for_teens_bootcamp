## 1. Arithmetic Operators

### Concept Explanation

**Arithmetic operators** are symbols that tell Python to perform a mathematical calculation — just like the buttons on a calculator.

| Operator | Symbol | Example | Result |
|----------|--------|---------|--------|
| Addition | `+` | `10 + 5` | `15` |
| Subtraction | `-` | `10 - 5` | `5` |
| Multiplication | `*` | `10 * 5` | `50` |
| Division | `/` | `10 / 4` | `2.5` |
| Floor Division | `//` | `10 // 4` | `2` (drops the decimal) |
| Modulus (Remainder) | `%` | `10 % 3` | `1` (leftover after dividing) |
| Exponent (Power) | `**` | `2 ** 3` | `8` (2 × 2 × 2) |

> 💡 **Floor division (`//`)** is useful when you need a whole number answer — like splitting chapatis equally among friends with no fractions!
>
> 💡 **Modulus (`%`)** tells you the *remainder*. For example, `10 % 3 = 1` because 3 goes into 10 three times, leaving 1 over.

````python
# Example 1: Calculating a shopping bill
# Scenario: Wanjiku buys items from Naivas supermarket.

unga_price    = 210    # price of 2 kg unga in Shillings
sukuma_price  = 30     # price of a bunch of sukuma wiki
chapati_price = 15     # price of one chapati
chapatis      = 4      # number of chapatis bought

# Addition: add up the items
total_bill = unga_price + sukuma_price + (chapati_price * chapatis)

# Subtraction: calculate change from Ksh 500
cash_given = 500
change     = cash_given - total_bill

print(f"Total bill  : Ksh {total_bill}")
print(f"Cash given  : Ksh {cash_given}")
print(f"Change      : Ksh {change}")
````

````python
# Example 2: Splitting mandazis equally (floor division and modulus)
# Scenario: Otieno brings 25 mandazis to share equally among 4 friends.

total_mandazis = 25
num_friends    = 4

# Floor division — how many whole mandazis each friend gets
each_gets  = total_mandazis // num_friends

# Modulus — how many are left over
left_over  = total_mandazis % num_friends

print(f"Each friend gets : {each_gets} mandazis")
print(f"Left over        : {left_over} mandazi(s)")
````

````python
# Example 3: Matatu fare and trip calculations
# Scenario: The fare from Kisumu to Nakuru is Ksh 350.
# Achieng wants to know the total cost for her family of 5.

fare_per_person = 350
family_size     = 5

# Multiplication: total cost for the family
total_fare = fare_per_person * family_size

# Exponent: just for fun — if the fare doubled every day for 3 days
doubled_fare = fare_per_person ** 2   # 350 to the power of 2

# Division: split cost between two parents
per_parent = total_fare / 2

print(f"Total fare for family  : Ksh {total_fare}")
print(f"Each parent contributes: Ksh {per_parent}")
````

## 2. Comparison Operators

### Concept Explanation

**Comparison operators** compare two values and always return either `True` or `False`. Think of them as asking Python a yes/no question:

> *"Is Kamau's score greater than the pass mark?"* → `True` or `False`

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `10 > 7` | `True` |
| `<` | Less than | `3 < 1` | `False` |
| `>=` | Greater than or equal to | `5 >= 5` | `True` |
| `<=` | Less than or equal to | `4 <= 3` | `False` |

> ⚠️ **Common mistake:** `=` *assigns* a value (`age = 16`), while `==` *compares* two values (`age == 16`). They are **not** the same!

````python
# Example 1: Comparing exam scores
# Scenario: Check whether Amina passed her exam. Pass mark is 50.

amina_score = 67
pass_mark   = 50

# Each comparison returns True or False
print("Did Amina pass?          ", amina_score >= pass_mark)   # True
print("Did she score exactly 50?", amina_score == pass_mark)   # False
print("Did she fail?            ", amina_score < pass_mark)    # False
````

````python
# Example 2: Comparing prices at two shops
# Scenario: Unga costs Ksh 210 at Naivas and Ksh 225 at the local duka.

naivas_price = 210
duka_price   = 225

print("Is Naivas cheaper?      ", naivas_price < duka_price)    # True
print("Are prices the same?    ", naivas_price == duka_price)   # False
print("Is the duka more expensive?", duka_price > naivas_price) # True
````

````python
# Example 3: Checking age for an event
# Scenario: A coding bootcamp only accepts students aged 13 and above.

odhiambo_age = 15
chebet_age   = 11
min_age      = 13

print("Can Odhiambo join?", odhiambo_age >= min_age)   # True
print("Can Chebet join?  ", chebet_age >= min_age)     # False
print("Same age?         ", odhiambo_age == chebet_age) # False
````

## 3. Logical Operators

### Concept Explanation

**Logical operators** let you combine multiple conditions into one. Instead of asking one yes/no question, you can ask *two or more at once*.

There are three logical operators:

| Operator | Meaning | Result is `True` when... |
|----------|---------|-------------------------|
| `and` | Both must be true | **Both** conditions are `True` |
| `or` | At least one must be true | **At least one** condition is `True` |
| `not` | Flip the result | The condition is `False` |

> 🎯 **Real-life analogy:**
> - **`and`** → To board a flight, you need *both* a ticket **and** a valid ID.
> - **`or`** → You can pay with M-Pesa **or** cash — either works.
> - **`not`** → The shop is **not** open means it is closed.

````python
# Example 1: Using 'and' — both conditions must be True
# Scenario: To join the school debate team, a student must be
# at least 14 years old AND have a score above 60.

zawadi_age   = 15
zawadi_score = 72

# Both conditions must be True for 'and' to return True
can_join = zawadi_age >= 14 and zawadi_score > 60

print("Can Zawadi join the debate team?", can_join)   # True
````

````python
# Example 2: Using 'or' — at least one condition must be True
# Scenario: A student gets a prize if they score above 90
# OR have perfect attendance (all 5 days present).

kamau_score      = 85
kamau_attendance = 5   # days present out of 5

# Only one needs to be True for 'or' to return True
gets_prize = kamau_score > 90 or kamau_attendance == 5

print("Does Kamau get a prize?", gets_prize)   # True (attendance is perfect)
````

````python
# Example 3: Using 'not' — flips True to False and vice versa
# Scenario: Check if the school library is closed.

library_open = True

print("Is the library open?  ", library_open)          # True
print("Is the library closed?", not library_open)      # False

# Combining all three
age     = 16
has_id  = False

# Must be 16+ AND have an ID, and NOT be banned
is_banned   = False
can_enter   = age >= 16 and has_id and not is_banned

print("Can the student enter?", can_enter)   # False — no ID
````

## 4. Conditional Statements — `if`, `elif`, `else`

### Concept Explanation

**Conditional statements** let your program make decisions — like a traffic light that tells you when to stop, slow down, or go.

```
🔴 Red light  → STOP  (if condition is True)
🟡 Amber light → SLOW (elif another condition is True)
🟢 Green light → GO   (else — none of the above)
```

**Syntax:**
```python
if condition:
    # code runs when condition is True
elif another_condition:
    # code runs when first is False but this is True
else:
    # code runs when ALL conditions above are False
```

> 🔑 **Key rules:**
> - Always end the `if`/`elif`/`else` line with a **colon `:`**
> - The code inside must be **indented** (4 spaces or 1 Tab)
> - `elif` and `else` are optional — you can have just an `if`

````python
# Example 1: Simple if-else — pass or fail
# Scenario: Check if Baraka passed his exam. Pass mark is 50.

baraka_score = 63
pass_mark    = 50

if baraka_score >= pass_mark:
    print("Baraka passed! Well done.")   # runs when score >= 50
else:
    print("Baraka did not pass. Keep going!")
````

````python
# Example 2: if-elif-else — grading system
# Scenario: Assign a letter grade to Zawadi's score.

zawadi_score = 78

if zawadi_score >= 80:
    grade = "A"          # 80 and above
elif zawadi_score >= 60:
    grade = "B"          # 60 to 79
elif zawadi_score >= 40:
    grade = "C"          # 40 to 59
else:
    grade = "D"          # below 40

# Only ONE block runs — the first condition that is True
print(f"Zawadi's score: {zawadi_score}")
print(f"Grade         : {grade}")
````

````python
# Example 3: Interactive — matatu ticket pricing
# Scenario: Matatu fares vary by age:
#   Children under 5  → free
#   Age 5 to 17       → Ksh 30
#   Adults 18+        → Ksh 50

age = int(input("Enter passenger age: "))

if age < 5:
    fare = 0
    category = "Child (free)"
elif age <= 17:
    fare = 30
    category = "Youth"
else:
    fare = 50
    category = "Adult"

print(f"Category : {category}")
print(f"Fare     : Ksh {fare}")
````

## 5. Loops — `for` Loop

### Concept Explanation

A **`for` loop** repeats a block of code for **each item in a sequence**. Instead of writing the same line 10 times, you write it once inside a loop.

> 🎯 **Real-life analogy:** The teacher calls out each student's name from the register — one by one, in order. That's a `for` loop!

**Syntax:**
```python
for variable in sequence:
    # code that runs for each item
```

- **`sequence`** can be a list, a string, or a `range()`.
- **`range(n)`** generates numbers from `0` up to (but not including) `n`.
- **`range(start, stop)`** generates numbers from `start` up to (but not including) `stop`.
- The loop variable takes the value of each item one at a time.

````python
# Example 1: Loop through a list of names
# Scenario: Print a welcome message for each student in the class.

students = ["Achieng", "Mwangi", "Zawadi", "Odhiambo", "Fatuma"]

# The loop goes through the list and runs the code for each name
for student in students:
    print(f"Welcome, {student}! Glad you're here.")
````

````python
# Example 2: Using range() to count and accumulate
# Scenario: Chebet saves Ksh 50 every day for 7 days. Show her savings each day.

daily_savings = 50
total_savings = 0

# range(1, 8) gives us the numbers 1, 2, 3, 4, 5, 6, 7
for day in range(1, 8):
    total_savings = total_savings + daily_savings   # add today's savings
    print(f"Day {day}: Total savings = Ksh {total_savings}")
````

````python
# Example 3: Loop through a list and apply logic inside
# Scenario: A list of exam scores — print a pass/fail message for each student.

results = [
    ("Baraka",  72),
    ("Njeri",   45),
    ("Kamau",   58),
    ("Atieno",  31),
    ("Muthoni", 89)
]

pass_mark = 50

for name, score in results:   # unpack each pair (name, score)
    if score >= pass_mark:
        status = "PASS ✓"
    else:
        status = "FAIL ✗"
    print(f"{name:<10} {score}   {status}")
````

## 6. Loops — `while` Loop

### Concept Explanation

A **`while` loop** keeps repeating as long as a condition is `True`. It doesn't know in advance how many times it will run — it just keeps going until the condition becomes `False`.

> 🎯 **Real-life analogy:** You keep eating chapatis *while* you're still hungry. Once you're full, you stop — you didn't know beforehand how many you'd eat!

**Syntax:**
```python
while condition:
    # code that keeps running while condition is True
    # make sure something changes so the condition eventually becomes False!
```

> ⚠️ **Infinite loop warning:** If the condition *never* becomes `False`, the loop runs forever and freezes your program! Always make sure something inside the loop changes the condition.

````python
# Example 1: Countdown — like a race starting signal
# Scenario: Count down from 5 before the Kisumu athletics race begins.

countdown = 5   # start at 5

while countdown > 0:            # keep going while countdown is above 0
    print(f"  {countdown}...")
    countdown = countdown - 1   # subtract 1 each time (this stops the loop eventually)

print("Go! 🏃")
````

````python
# Example 2: Saving towards a goal
# Scenario: Ochieng saves Ksh 200 a week.
# How many weeks until he has saved at least Ksh 1000 for a new football?

savings      = 0      # start with nothing
weekly_save  = 200    # saves this much every week
goal         = 1000   # target amount
weeks        = 0      # week counter

while savings < goal:               # keep saving until goal is reached
    savings = savings + weekly_save
    weeks   = weeks + 1
    print(f"Week {weeks}: Ksh {savings} saved")

print(f"\nOchieng reached his goal in {weeks} weeks!")
````

````python
# Example 3: Input validation — keep asking until a valid answer is given
# Scenario: A quiz asks a question and keeps prompting until the right answer.

correct_answer = "Nairobi"
answer = ""   # start with an empty answer

# Keep looping until the student types the right answer
while answer != correct_answer:
    answer = input("What is the capital city of this country? ")
    if answer != correct_answer:
        print("Not quite — try again!")

print("Correct! Well done! 🎉")
````
