## 1. Output — Displaying Information

### Concept Explanation

**Output** is how a program communicates results or messages to the user. In Python, the built-in `print()` function is used to display information on the screen.

**Syntax:**
```python
print(value)
```

- You can print text (called a **string**) by wrapping it in quotes.
- You can print numbers directly without quotes.
- You can print multiple values by separating them with commas — Python will add a space between them automatically.
- The `end` parameter controls what is printed at the end of a line (default is a newline `\n`).
- The `sep` parameter controls the separator between multiple values (default is a space).

````python
# Example 1: Basic print statements

# Printing a simple greeting (text must be wrapped in quotes)
print("Hello, World!")

# Printing a number (no quotes needed for numbers)
print(47)   # total number of counties in the country

# Printing multiple values separated by commas
# Python automatically adds a space between them
print("My name is", "Kamau")
````

````python
# Example 2: Customising print output with sep and end

# Using sep to change the separator between values
# Format a date as DD/MM/YYYY
print(16, 3, 2026, sep="/")    # Output: 16/3/2026

# Using end to stay on the same line instead of moving to a new line
print("Nairobi", end=" -> ")
print("Mombasa")               # Output: Nairobi -> Mombasa

# Printing an empty line (useful for spacing output)
print()
print("Have a safe trip!")     # blank line appears above this
````

````python
# Example 3: Printing with f-strings (formatted strings)
# f-strings let you embed variable values directly inside a string using {}

student_name = "Achieng"   # variable holding a student's name
score        = 87           # variable holding the student's score

# The f before the quote makes it an f-string
# Anything inside {} is replaced with the variable's value
print(f"Name: {student_name}")
print(f"{student_name} scored {score} out of 100.")
````

## 2. User Input — Making Programs Interactive

### Concept Explanation

**Input** allows a program to receive data from the user while it is running, making the program interactive.

In Python, the `input()` function is used to collect user input.

**Syntax:**
```python
variable = input("Prompt message: ")
```

**Important rules:**
- `input()` always returns the data as a **string**, even if the user types a number.
- To use the input as a number, you must **convert** it using `int()` (for whole numbers) or `float()` (for decimal numbers).
- The prompt message inside `input()` is displayed to the user so they know what to type.

````python
# Example 1: Reading a name from the user

# input() pauses the program and waits for the user to type something
# Whatever the user types is stored in the variable 'name'
name = input("Enter your name: ")

# Use the stored value to personalise the greeting
print("Hello,", name + "! Welcome to class.")
````

````python
# Example 2: Reading a number and performing a calculation

# Scenario: A matatu charges Ksh 50 per trip.
# Ask the user how many trips they made and calculate the total fare.

# input() always gives a string, so we wrap it with int()
# to convert it to a whole number we can do maths with
trips = int(input("How many trips did you take today? "))

fare_per_trip = 50    # fare per trip in Shillings
total_fare    = trips * fare_per_trip

print(f"Total fare for today: Ksh {total_fare}")
````

````python
# Example 3: Reading two numbers and adding them

# Scenario: Mama Njeri wants to add the prices of two items at her shop.

# float() converts input to a decimal number — useful for prices
sugar_price = float(input("Price of sugar (Ksh): "))
maize_price = float(input("Price of maize flour (Ksh): "))

# Calculate the total bill
total = sugar_price + maize_price

print(f"Total bill: Ksh {total}")
````

## 3. Variables — Storing Information

### Concept Explanation

A **variable** is a named container that stores a value in a program's memory. Think of it like a labelled tin (debe) in a kitchen — you give it a name, put something inside, and take it out whenever you need it.

**Syntax:**
```python
variable_name = value
```

**Naming rules:**
- Must start with a **letter** or underscore `_` (not a number).
- Can contain letters, numbers, and underscores.
- **Case-sensitive**: `Alama` and `alama` are two different variables.
- Cannot be a Python keyword (e.g., `print`, `if`, `for`).
- Use descriptive names that explain what the variable holds (`jina_mwanafunzi` is better than `x`).

The `=` sign is the **assignment operator** — it stores the value on the right into the variable on the left.

````python
# Example 1: Creating and using variables

# Assigning values to variables about a student
student_name = "Mwangi"   # stores the student's name
age          = 20          # stores the student's age
gpa          = 3.6         # stores the GPA (decimal number)

# Using the variables in a print statement
print(student_name, "is", age, "years old.")
print(f"GPA: {gpa}")
````

````python
# Example 2: Updating a variable's value
# A variable can be reassigned — its value changes to the new one

# Scenario: Chebet is saving money for school fees.
savings = 5000           # starting savings in Shillings
print("Initial savings: Ksh", savings)

savings = savings + 2000  # Chebet added Ksh 2000; same as: savings += 2000
print("Savings after top-up: Ksh", savings)

# The old value (5000) is gone; savings now holds 7000
````

````python
# Example 3: Swapping two variable values
# Scenario: Swap the bench assignments of two students.

seat_a = "Odhiambo"
seat_b = "Njeri"

print("Before swap — Seat A:", seat_a, "| Seat B:", seat_b)

# Python allows swapping in one line using tuple unpacking
seat_a, seat_b = seat_b, seat_a

print("After swap  — Seat A:", seat_a, "| Seat B:", seat_b)
````

## 4. Data Types — Kinds of Information

### Concept Explanation

Every value in Python has a **data type** that tells Python what kind of data it is and what operations are allowed on it.

| Data Type | Python keyword | Example | Description |
|-----------|---------------|---------|-------------|
| String    | `str`         | `"Nairobi"` | Text — any characters in quotes |
| Integer   | `int`         | `47`        | Whole numbers (no decimal point) |
| Float     | `float`       | `210.50`    | Decimal numbers |
| Boolean   | `bool`        | `True` / `False` | Represents yes/no, on/off |

Use `type()` to check the data type of any value or variable.

**Type conversion** (casting) allows you to change a value from one type to another:
- `int("5")` → `5`
- `float(3)` → `3.0`
- `str(100)` → `"100"`
- `bool(0)` → `False`

````python
# Example 1: The four basic data types

city         = "Kisumu"    # str   — text describing a city name
num_counties = 47          # int   — the country has 47 counties (whole number)
unga_price   = 210.50      # float — price of a 2 kg bag of unga flour (decimal)
is_capital   = True        # bool  — True if the place is a capital city

# type() returns the data type of a value
print(type(city))          # <class 'str'>
print(type(num_counties))  # <class 'int'>
print(type(unga_price))    # <class 'float'>
print(type(is_capital))    # <class 'bool'>
````

````python
# Example 2: Why data types matter — string vs number

# Scenario: Kamau and Atieno each send money via M-Pesa.
# What happens when their amounts are stored as text instead of numbers?

# When + is used with strings it joins them (concatenation)
kamau_amount  = "500"
atieno_amount = "300"
print("String  +:", kamau_amount + atieno_amount)    # "500300" — joined as text! Wrong!

# When + is used with integers it adds them (arithmetic)
kamau_amount  = 500
atieno_amount = 300
print("Integer +:", kamau_amount + atieno_amount)    # 800     — correct total

# This is why we convert input() strings to int/float before doing maths
````

````python
# Example 3: Type conversion (casting)

# Convert string to integer — useful after reading input()
# Scenario: A student's score was accidentally stored as text
score_text   = "85"             # score stored as text (e.g., from input)
score_number = int(score_text)  # convert to integer so we can do maths
bonus        = 5
print("Score with bonus:", score_number + bonus)   # 90 — maths works now

# Convert integer to string — useful when building messages
chapati_price = 15
message = "One chapati costs Ksh " + str(chapati_price) + " only!"
print(message)

# Boolean values: 0 is False, everything else is True
print(bool(0))      # False — e.g., student paid Ksh 0 fees → not paid
print(bool(5000))   # True  — student paid Ksh 5000 fees → paid
print(bool(""))     # False — empty name field → no name entered
````

## 5. Comments — Explaining Your Code

### Concept Explanation

**Comments** are notes written inside code to explain what it does. Python **ignores comments** when running the program — they are only there for humans to read.

Good comments make code easier to understand, maintain, and debug.

**Single-line comment:** starts with `#`
```python
# This is a comment
price = 50  # price in Shillings
```

**Multi-line comment (docstring):** wrapped in triple quotes `"""` or `'''`
```python
"""
This is a multi-line comment.
It can span several lines.
"""
```

**Best practices:**
- Comment the *why*, not the *what* (the code already shows what).
- Keep comments up-to-date when you change the code.
- Avoid over-commenting obvious things like `price = 50  # set price to 50`.

````python
# Example 1: Single-line comments

# This program calculates Amina's average exam score

mid_score = 68    # mid-semester exam score out of 100
end_score = 74    # end-semester exam score out of 100

# Average the two scores to get the final grade
average = (mid_score + end_score) / 2

print("Amina's average score:", average)   # display the result
````

````python
# Example 2: Multi-line comment / docstring

"""
Program : Matatu Fare Calculator
Author  : Student Name
Date    : March 2026

This program asks the user for the trip distance and fare rate per km,
then calculates and displays the total fare to be paid.
"""

distance    = float(input("Trip distance (km): "))
rate_per_km = float(input("Fare rate per km (Ksh): "))

# Calculate total fare: distance multiplied by rate per km
total_fare = distance * rate_per_km

print(f"Total fare to pay: Ksh {total_fare:.2f}")
````

````python
# Example 3: Using comments to temporarily disable code
# This technique is useful during testing and debugging

sukuma_price = 30   # current price of sukuma wiki per bunch

# print("Debug: sukuma_price =", sukuma_price)  <- commented out during testing
# The line above won't run, but we keep it in case we need it later

print("Price of sukuma wiki: Ksh", sukuma_price)
````
