## 1. Introduction to File Systems

```
💻 Computer (Root)
└── 📁 Users
```

| **Root** | The very top of the file system | `C:\` on Windows, `/` on Mac/Linux |

| **Extension** | The letters after the dot — tells you the file type | `.txt`, `.py`, `.jpg`, `.csv` |

## 2. Understanding File Paths

### Concept Explanation

A **file path** is the full address of a file or folder on your computer — like a GPS coordinate that tells the computer exactly where to find something.

**Parts of a file path:**
```
C:\Users\Wanjiru\Documents\School\results.txt
│    │           │           │      │
Drive Root       Folder      Sub-   Filename
                             folder  (with extension)
```

**Path separators — the slash between folders:**

| Operating System | Separator | Example |
|-----------------|-----------|--------|
| Windows | `\` (backslash) | `C:\Users\Wanjiru\notes.txt` |
| Mac / Linux | `/` (forward slash) | `/home/wanjiru/notes.txt` |
| Python (safe on all) | `/` or `os.path.join()` | `os.path.join("Users", "Wanjiru", "notes.txt")` |



> 💡 **Best practice:** Use `os.path.join()` in your Python code instead of typing slashes manually. It automatically uses the correct separator for whatever computer your code runs on — so your program works on Windows, Mac, and Linux!

````python
# Example 1: Building file paths safely with os.path.join()
# This avoids the Windows vs Mac slash problem.

import os

# os.path.join() joins folder names using the correct separator for this computer
path1 = os.path.join("Documents", "School", "results.txt")
path2 = os.path.join("Users", "Odhiambo", "Projects", "bootcamp.py")

print("Path 1:", path1)
print("Path 2:", path2)
````

````python
# Example 2: Breaking a path into its parts
# os.path gives tools to inspect parts of any path.

import os

filepath = os.path.join("Documents", "School", "achieng_notes.txt")

# Get just the filename from the path
filename = os.path.basename(filepath)

# Get just the folder part
folder = os.path.dirname(filepath)

# Split filename into name and extension
name, extension = os.path.splitext(filename)

print("Full path :", filepath)
print("File name :", filename)
print("Folder    :", folder)
print("Name only :", name)
print("Extension :", extension)
````

````python
# Example 3: Getting the file size and checking if something is a file or folder

import os

# Check each item in the current directory
print(f"{'Name':<35} {'Type':<10}")
print("-" * 45)

for item in os.listdir():
    # os.path.isfile() returns True if the item is a file
    # os.path.isdir() returns True if the item is a directory/folder
    if os.path.isfile(item):
        item_type = "File"
    elif os.path.isdir(item):
        item_type = "Folder"
    else:
        item_type = "Other"

    print(f"{item:<35} {item_type:<10}")
````

## 3. Absolute vs Relative Paths

### Concept Explanation

There are two ways to describe where a file is:

---

### 📍 Absolute Path
An **absolute path** gives the *complete* address of a file, starting from the very root of the file system. It works no matter where you are on the computer.

```
Windows : C:\Users\Zawadi\Documents\School\notes.txt
Mac/Linux: /home/zawadi/Documents/School/notes.txt
```

### 📂 Relative Path
A **relative path** gives the address of a file *relative to where you currently are*. It only works correctly from that starting location.

```
If you are already in:  C:\Users\Zawadi\Documents\
Relative path to notes: School\notes.txt
```

**Special relative path symbols:**

| Symbol | Meaning | Example |
|--------|---------|--------|
| `.` | Current folder | `./notes.txt` = notes.txt in this folder |
| `..` | One folder UP | `../photos/` = go up, then into photos |



> 🎯 **Analogy:** Giving directions to Naivas supermarket.
> - **Absolute:** "Go to Moi Avenue, Nairobi CBD, opposite the post office."
> - **Relative:** "Turn left out of school, walk 200m, it's on the right."
> The relative directions only make sense if you know where the school is!

````python
# Example 1: Getting the absolute path of the current working directory

import os

# os.getcwd() returns the absolute path of where Python is currently running
absolute_cwd = os.getcwd()
print("Absolute path of current folder:")
print(" ", absolute_cwd)

# os.path.abspath() converts any path (relative or absolute) into a full absolute path
relative = "sample_files"
absolute = os.path.abspath(relative)
print("\nRelative path :", relative)
print("Absolute path :", absolute)
````

````python
# Example 2: Understanding the '..' (go up one folder) notation

import os

current = os.getcwd()
print("Current folder     :", current)

# '..' means 'go up one level' — to the parent folder
one_up  = os.path.abspath("..")
print("One folder up      :", one_up)

# '../..' goes up two levels
two_up  = os.path.abspath("../..")
print("Two folders up     :", two_up)
````

````python
# Example 3: Relative vs absolute paths in action
# Scenario: Check whether two paths point to the same file.

import os

# These might look different but could be the same location
relative_path = "."
absolute_path = os.getcwd()

# Convert both to absolute paths and compare
abs_relative = os.path.abspath(relative_path)
abs_absolute = os.path.abspath(absolute_path)

print("Path A (relative '.'):", abs_relative)
print("Path B (absolute cwd):", abs_absolute)
print("Are they the same?   :", abs_relative == abs_absolute)
````

## 4. Working with Directories and Files

### Concept Explanation

Python's `os` module lets you **create, rename, and delete** folders and files directly from your code — just like right-clicking in File Explorer, but using Python.

**Useful `os` functions:**

| Function | What it does |
|----------|--------------|
| `os.mkdir("folder")` | Create a **single** new folder |
| `os.makedirs("a/b/c")` | Create a folder **and all parent folders** needed |
| `os.rename("old", "new")` | Rename a file or folder |
| `os.remove("file.txt")` | Delete a file |
| `os.rmdir("folder")` | Delete an **empty** folder |
| `os.path.exists("path")` | Check if a file or folder exists |
| `os.path.isfile("path")` | Check if a path is a file |
| `os.path.isdir("path")` | Check if a path is a directory |

> ⚠️ Always check if a folder **already exists** before trying to create it — otherwise Python throws an error! Use `if not os.path.exists(folder):` before `os.mkdir()`.

````python
# Example 1: Create a folder to hold our practice files
# We will use this 'sample_files' folder in all the file reading/writing examples below.

import os

folder_name = "sample_files"

# Only create the folder if it doesn't already exist
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"Folder '{folder_name}' created successfully.")
else:
    print(f"Folder '{folder_name}' already exists — no action needed.")

print("Current directory contents:", os.listdir())
````

````python
# Example 2: Create a nested folder structure in one go with os.makedirs()
# Scenario: Set up a school project folder with sub-folders for each subject.

import os

# Create multiple nested folders at once
# exist_ok=True means: don't raise an error if the folder already exists
os.makedirs("sample_files/school/maths",   exist_ok=True)
os.makedirs("sample_files/school/english", exist_ok=True)
os.makedirs("sample_files/school/science", exist_ok=True)

print("Folders created inside sample_files/school/:")
for item in os.listdir("sample_files/school"):
    print(f"  📁 {item}")
````

````python
# Example 3: Rename a file and check what's inside a folder
# Scenario: First create a temporary file, then rename it.

import os

# Create a temporary empty file to practice renaming
old_name = "sample_files/temp_file.txt"
new_name = "sample_files/ochieng_notes.txt"

# Create the file first by opening it in write mode
open(old_name, "w").close()   # creates an empty file
print("Before rename:", os.listdir("sample_files"))

# Rename the file
os.rename(old_name, new_name)
print("After rename :", os.listdir("sample_files"))
````

## 5. Opening Files in Python

### Concept Explanation

To work with a file in Python, you must first **open** it using the built-in `open()` function. This is like picking up a book from the shelf before you can read or write in it.

```python
file = open("filename.txt", mode)
```

**File modes — what do you want to do with the file?**

| Mode | Symbol | What it does |
|------|--------|-------------|
| Read | `"r"` | Opens for reading. Error if file doesn't exist. |
| Write | `"w"` | Opens for writing. **Creates** the file if it doesn't exist. **Erases** existing content! |
| Append | `"a"` | Opens for writing **at the end**. Creates if needed. Keeps existing content. |
| Read + Write | `"r+"` | Opens for both reading and writing. |

**The `with` statement — the safest way to open files:**
```python
with open("filename.txt", "r") as file:
    content = file.read()
# File is automatically closed here — even if an error occurs!
```



> 💡 **Always use `with`!** It automatically closes the file when you're done, which prevents data loss and errors. Think of it like automatically putting the book back on the shelf.

````python
# Example 1: Create a text file using write mode ('w')
# This is like taking out a blank exercise book and writing the title.

filepath = "sample_files/mwangi_diary.txt"

# Open in write mode — creates the file if it doesn't exist
with open(filepath, "w") as diary:
    diary.write("Dear Diary,\n")                     # \n means new line
    diary.write("Today was a great day at bootcamp.\n")
    diary.write("I learned how to create files with Python!\n")

# After the 'with' block, the file is automatically closed
print(f"File '{filepath}' created successfully.")
print("Does it exist now?", os.path.exists(filepath))
````

````python
# Example 2: The difference between 'w' (overwrite) and 'a' (append)
# IMPORTANT: 'w' wipes the file clean before writing. 'a' adds to the end.

filepath = "sample_files/test_modes.txt"

# Step 1: Write initial content
with open(filepath, "w") as f:
    f.write("Line 1: Written with 'w' mode.\n")

# Step 2: Append — adds to the existing content
with open(filepath, "a") as f:
    f.write("Line 2: Added with 'a' mode.\n")

# Step 3: Write again with 'w' — THIS ERASES EVERYTHING!
with open(filepath, "w") as f:
    f.write("Line 3: Written with 'w' again — previous lines are gone!\n")

# Read back the final content
with open(filepath, "r") as f:
    print("Final file content:")
    print(f.read())
````

````python
# Example 3: Check if a file exists BEFORE opening it to avoid errors
# Trying to open a non-existent file in 'r' mode causes a FileNotFoundError.

import os

def safe_open(filepath):
    """Open a file only if it exists — otherwise print a helpful message."""
    if os.path.exists(filepath):
        with open(filepath, "r") as f:
            print(f"Contents of '{filepath}':")
            print(f.read())
    else:
        print(f"File not found: '{filepath}'")
        print("Please check the file name and path.")

safe_open("sample_files/mwangi_diary.txt")   # exists
print()
safe_open("sample_files/missing_file.txt")   # does not exist
````

## 6. Reading from Text Files

### Concept Explanation

Once a file is open in read mode (`"r"`), Python gives you three main ways to read its content:

| Method | What it returns | Best for |
|--------|----------------|----------|
| `.read()` | The **entire file** as one big string | Small files |
| `.readline()` | **One line** at a time (moves a bookmark forward) | Processing line by line manually |
| `.readlines()` | A **list** where each item is one line | When you need to work with all lines as a list |

**Looping directly over a file** is the most common and efficient technique:
```python
with open("file.txt", "r") as f:
    for line in f:        # Python reads one line at a time automatically
        print(line)
```

> 💡 Each line in a text file ends with `\n` (a newline character). Use `.strip()` to remove it:
> ```python
> line.strip()   # removes spaces and \n from both ends
> ```

````python
# First, create a file with student names to read from
filepath = "sample_files/students.txt"

with open(filepath, "w") as f:
    f.write("Achieng\n")
    f.write("Baraka\n")
    f.write("Chebet\n")
    f.write("Fatuma\n")
    f.write("Kamau\n")

print("Sample file created:", filepath)
````

````python
# Example 1: read() — read the entire file at once

filepath = "sample_files/students.txt"

with open(filepath, "r") as f:
    all_content = f.read()   # returns one big string with all text

print("--- Full file content ---")
print(all_content)
print("Type returned:", type(all_content))
````

````python
# Example 2: readlines() — get all lines as a list

filepath = "sample_files/students.txt"

with open(filepath, "r") as f:
    lines = f.readlines()   # returns a list like ['Achieng\n', 'Baraka\n', ...]

print("Raw list          :", lines)
print("Number of students:", len(lines))

print("\nCleaned up (using .strip()):")
for i, line in enumerate(lines, start=1):
    name = line.strip()    # removes the \n at the end of each line
    print(f"  {i}. {name}")
````

````python
# Example 3: Loop directly over a file — the most efficient approach
# Scenario: Read a results file and print a pass/fail message for each student.

# First create the results file
results_path = "sample_files/results.txt"
with open(results_path, "w") as f:
    f.write("Zawadi,78\n")
    f.write("Odhiambo,42\n")
    f.write("Muthoni,85\n")
    f.write("Otieno,55\n")

# Now read and process it line by line
print("STUDENT RESULTS")
print("-" * 30)

with open(results_path, "r") as f:
    for line in f:
        line  = line.strip()          # remove \n
        name, score = line.split(",") # split "Zawadi,78" into name and score
        score = int(score)            # convert score from string to integer

        status = "PASS" if score >= 50 else "FAIL"
        print(f"{name:<12} {score}   {status}")
````

## 7. Writing to Text Files

### Concept Explanation

Writing to files lets your program **save data permanently** — even after the program closes, the data stays in the file. This is how apps save user information, game scores, and settings.

**Two main writing methods:**

| Method | What it does |
|--------|--------------|
| `.write(string)` | Writes a single string — you control everything, including `\n` |
| `.writelines(list)` | Writes a list of strings — each string must include its own `\n` |

**Remember the modes:**
- `"w"` → Starts fresh (old content is **deleted**)
- `"a"` → Adds to the end (old content is **kept**)



> ⚠️ **Common mistake:** Forgetting `\n` at the end of each `.write()` call. Without it, all your text runs together on one line in the file!

````python
# Example 1: Writing student scores to a file from a list
# Scenario: Save the term results to a file so they can be retrieved later.

results = [
    ("Achieng",  88),
    ("Baraka",   55),
    ("Chebet",   72),
    ("Fatuma",   43),
    ("Ochieng",  91)
]

filepath = "sample_files/term_results.txt"

with open(filepath, "w") as f:
    f.write("NAME,SCORE\n")               # header line
    for name, score in results:
        f.write(f"{name},{score}\n")       # write each result as a CSV line

print(f"Results saved to '{filepath}'.")

# Verify by reading it back
with open(filepath, "r") as f:
    print(f.read())
````

````python
# Example 2: writelines() — write a whole list at once
# Scenario: Save a canteen menu to a file.

menu_items = [
    "Ugali and sukuma wiki - Ksh 60\n",
    "Chapati - Ksh 20\n",
    "Pilau - Ksh 80\n",
    "Chai - Ksh 20\n",
    "Mandazi - Ksh 10\n"
]

filepath = "sample_files/canteen_menu.txt"

with open(filepath, "w") as f:
    f.write("=== TODAY'S CANTEEN MENU ===\n")
    f.writelines(menu_items)              # writes all items from the list
    f.write("============================\n")

print("Menu saved. Reading it back:")
with open(filepath, "r") as f:
    print(f.read())
````

````python
# Example 3: Appending new entries to an existing file
# Scenario: A new student (Wanjiru) joins mid-term — add her to the results file.

filepath = "sample_files/term_results.txt"

print("Before appending:")
with open(filepath, "r") as f:
    print(f.read())

# Append the new student — 'a' mode keeps existing content
with open(filepath, "a") as f:
    f.write("Wanjiru,79\n")

print("After appending Wanjiru:")
with open(filepath, "r") as f:
    print(f.read())
````

## 8. Closing Files and Best Practices

### Concept Explanation

When you open a file, Python keeps a connection to it. If you forget to **close** the file, you risk:
- **Data loss** — some content may not be saved to disk
- **File corruption** — the file may become unreadable
- **Resource leaks** — your program holds onto memory it doesn't need

**Two ways to close files:**

| Approach | Code | Risk |
|----------|------|------|
| Manual close | `f = open(...)` then `f.close()` | If an error happens before `f.close()`, the file stays open |
| `with` statement ✅ | `with open(...) as f:` | Automatically closes even if an error occurs — **always preferred** |

**Best practices checklist:**
- ✅ Always use the `with` statement
- ✅ Use `os.path.exists()` before opening a file for reading
- ✅ Use relative paths so your code works on any computer
- ✅ Use `"a"` mode to add to a file, not `"w"` (which deletes existing content)
- ✅ Use `.strip()` when reading lines to remove unwanted whitespace and `\n`
- ✅ Handle errors with `try / except` for production code

````python
# Example 1: The wrong way vs the right way to open files

import os

filepath = "sample_files/mwangi_diary.txt"

# ❌ Wrong way — if an error occurs between open() and close(), file stays open
f = open(filepath, "r")
content = f.read()
f.close()   # must remember to call this!
print("Read with manual close — OK, but risky.")

# ✅ Right way — 'with' guarantees the file is always closed
with open(filepath, "r") as f:
    content = f.read()
# File is closed here automatically — no need to call f.close()
print("Read with 'with' statement — safe and clean.")
````

````python
# Example 2: Handling missing files gracefully with try/except
# Scenario: Try to open a file — if it doesn't exist, show a friendly message.

def read_file_safely(filepath):
    """Read a file and return its content, or an error message if not found."""
    try:
        with open(filepath, "r") as f:
            return f.read()
    except FileNotFoundError:
        return f"Sorry, the file '{filepath}' was not found."

# Test with a real file and a missing file
print(read_file_safely("sample_files/students.txt"))
print()
print(read_file_safely("sample_files/does_not_exist.txt"))
````

````python
# Example 3: A complete, best-practice file handling function
# Scenario: Save bootcamp attendance records safely.

import os

def save_attendance(filename, records):
    """
    Saves a list of (name, present) attendance records to a file.
    Uses append mode so existing records are not lost.
    """
    # Ensure the folder exists before trying to write
    folder = os.path.dirname(filename)
    if folder and not os.path.exists(folder):
        os.makedirs(folder)

    with open(filename, "a") as f:        # 'a' keeps existing records
        for name, present in records:
            status = "Present" if present else "Absent"
            f.write(f"{name},{status}\n")

    print(f"{len(records)} record(s) saved to '{filename}'.")

# Daily attendance for the bootcamp
today_attendance = [
    ("Achieng",  True),
    ("Kamau",    True),
    ("Zawadi",   False),
    ("Odhiambo", True)
]

save_attendance("sample_files/attendance.txt", today_attendance)

# Read back to confirm
with open("sample_files/attendance.txt", "r") as f:
    print("\nAttendance file contents:")
    print(f.read())
````

| Concept | Key Tool / Syntax | Example |
