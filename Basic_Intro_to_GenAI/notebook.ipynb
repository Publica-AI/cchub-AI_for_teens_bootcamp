## 1.2 Introduction to Large Language Models (LLMs)

### Concept Explanation

A **Large Language Model (LLM)** is the brain behind modern AI chatbots.

Breaking the name down:
- **Large** — trained on a HUGE amount of text (billions of books, websites, articles)
- **Language** — it works with human language (English, Swahili, French, etc.)
- **Model** — it is a mathematical program that has learned patterns from all that text

### A Simple Analogy

Imagine a student called **Zawadi** who reads every single book in the **University of Nairobi library** — all the textbooks, novels, newspapers, websites, everything. After all that reading, Zawadi becomes incredibly knowledgeable and can answer almost any question.

An LLM is like Zawadi — except it has "read" hundreds of times more than any human ever could!

### What Makes an LLM "Large"?

LLMs have **billions of parameters** — numbers that store what the model has learned.

| Model | Parameters (billions) | Made by |
|-------|----------------------|---------|
| GPT-4o | ~200 | OpenAI |
| Llama 3 | 8 – 70 | Meta |
| Gemma | 2 – 9 | Google |

That is like having billions of tiny "memory cells" storing knowledge!

---
## 1.3 How Do LLMs Learn?

### Concept Explanation

LLMs learn through a process called **training**. Here is how it works in simple steps:

**Step 1 — Collect Data**
Engineers collect massive amounts of text from the internet: Wikipedia, books, news articles, social media, code, etc.

**Step 2 — Learn to Predict**
The model is trained by playing a guessing game over and over:
> "The capital of Kenya is _____"

The model guesses the next word. If it guesses wrong, it adjusts its internal numbers. If it guesses right, it keeps those settings.

**Step 3 — Repeat Billions of Times**
This guessing-and-adjusting happens billions of times until the model becomes very good at predicting text.

**Step 4 — Fine-tuning**
Human trainers then teach the model to be helpful, harmless, and honest — like coaching a student before an important exam.

### Analogy
Think of it like learning to cook **ugali**:
1. You watch someone cook it (collect data)
2. You try yourself and it comes out lumpy (bad prediction)
3. You adjust — add more water, stir differently
4. After many tries, your ugali is perfect!

LLMs do the same — but with words instead of ugali, and billions of tries instead of a few!

LLMs are incredibly versatile. Here are some things they can do:

| Task | Example |

> 💡 **Golden Rule**: LLMs are powerful assistants, not perfect oracles. Use them to help you think, not to think for you!

## 2.4 Working With OpenAI Endpoints

### Concept Explanation

An **endpoint** is a specific URL (web address) in the API that does a specific job.

OpenAI has several endpoints:

| Endpoint | What it does |
|----------|--------------|
| **Chat Completions** | Have a conversation with the AI (this is the main one we use!) |
| Image Generation | Create images from text descriptions |
| Audio Transcription | Convert speech to text |
| Embeddings | Convert text into numbers for search |

We will focus on the **Chat Completions** endpoint — it is what powers ChatGPT!

### How a Chat Message is Structured

When we talk to an LLM, messages follow a structure with **roles**:

| Role | Who Is It? | Example |
|------|-----------|--------|
| `system` | Instructions to the AI | "You are a helpful Kenyan tutor" |
| `user` | Your message to the AI | "What is the capital of Kenya?" |
| `assistant` | The AI's reply | "The capital of Kenya is Nairobi." |

Think of `system` as the rules given to a teacher before class, and `user` as a student raising their hand to ask a question.

## 2.5 Ask The LLM a Question (OpenAI)

### Concept Explanation

Now let us write our first Python program that talks to an AI!

> ⚠️ **You need an OpenAI API key for this section.**
> If you do not have one yet, skip ahead to Section 2.7 (Ollama) which is completely free!

````python
# Step 1: Import the OpenAI library
from openai import OpenAI

# Step 2: Set your API key
# Replace the string below with your actual API key
# NEVER share this key with anyone — treat it like a password!
api_key = "sk-your-api-key-here"   # <-- paste your key here

# Step 3: Create the OpenAI client (the 'bridge' to talk to OpenAI)
client = OpenAI(api_key=api_key)

print("OpenAI client ready!")
````

````python
# Step 4: Ask the AI a question!

# The question we want to ask
my_question = "What are 3 interesting facts about Lake Victoria?"

# Send the question to OpenAI
# We use 'gpt-4o-mini' — it is fast and cheaper than the full GPT-4o
response = client.chat.completions.create(
    model="gpt-4o-mini",      # which AI model to use
    messages=[
        {
            "role": "system",          # give the AI its role
            "content": "You are a helpful assistant for Kenyan students."
        },
        {
            "role": "user",            # our question
            "content": my_question
        }
    ]
)

# Extract and print the AI's reply
ai_reply = response.choices[0].message.content
print("AI says:")
print(ai_reply)
````

### ✏️ Practice — Multi-turn Conversation

A chatbot should remember what was said earlier in a conversation.
Look at the code below — it builds a list of messages that grows with each exchange.

Run it and have a 3-message conversation with the AI about **nyama choma** or any Kenyan food!

````python
# Practice — Multi-turn conversation with Ollama
# The AI will remember everything said in the conversation

import ollama

# Start the conversation history with a system message
conversation_history = [
    {
        "role": "system",
        "content": "You are a friendly food expert who loves Kenyan cuisine. Keep answers short and fun."
    }
]

# Have a 3-turn conversation
my_messages = [
    "What is nyama choma?",
    "How do you prepare it traditionally?",
    "Which region in Kenya is most famous for it?"
]

for message in my_messages:
    # Add our message to the conversation history
    conversation_history.append({"role": "user", "content": message})
    
    # Get the AI's response
    response = ollama.chat(model="llama3.2", messages=conversation_history)
    
    # Get the AI's reply
    ai_reply = response["message"]["content"]
    
    # Add the AI's reply to the history (so it 'remembers' it next time)
    conversation_history.append({"role": "assistant", "content": ai_reply})
    
    # Print the conversation
    print(f"You:  {message}")
    print(f"AI:   {ai_reply}")
    print("-" * 50)
````

## 3.3 Understanding the Chatbot Code

Let us walk through the main parts of the chatbot app so you understand how it works:

````python
# This is a simplified version of the chatbot to explain how it works.
# The actual working app is in 'ollama_chatbot_app.py'
# DO NOT run this cell — it only works as a .py file with Streamlit!

import streamlit as st   # Streamlit for the web interface
import ollama            # Ollama to talk to the AI

# --- Page Setup ---
# Set the title that appears in the browser tab
st.title("🤖 Msaidizi Wangu — My AI Assistant")
st.caption("Powered by Ollama | Running on your laptop!")

# --- Memory: Store Conversation History ---
# st.session_state remembers things between messages in the same session
# It is like the conversation_history list we used before!
if "messages" not in st.session_state:
    st.session_state.messages = [
        {
            "role": "system",
            "content": "You are Msaidizi, a helpful AI assistant for Kenyan students. Be friendly and clear."
        }
    ]

# --- Display Past Messages ---
# Loop through all messages and show them in the chat window
for msg in st.session_state.messages:
    if msg["role"] != "system":   # don't show the system message to users
        with st.chat_message(msg["role"]):  # show as 'user' or 'assistant' bubble
            st.markdown(msg["content"])

# --- Input Box ---
# st.chat_input creates the message box at the bottom of the page
user_input = st.chat_input("Type your message here...")

if user_input:  # if the user typed something and pressed Enter
    
    # Show the user's message in the chat
    with st.chat_message("user"):
        st.markdown(user_input)
    
    # Add the user's message to history
    st.session_state.messages.append({"role": "user", "content": user_input})
    
    # Get the AI's reply from Ollama
    response = ollama.chat(
        model="llama3.2",
        messages=st.session_state.messages
    )
    ai_reply = response["message"]["content"]
    
    # Show the AI's reply in the chat
    with st.chat_message("assistant"):
        st.markdown(ai_reply)
    
    # Add the AI's reply to history so it 'remembers'
    st.session_state.messages.append({"role": "assistant", "content": ai_reply})
````

## 3.4 Keeping Your API Key Safe When Deploying

### The Problem — Never Hardcode Your Key!

When you write `api_key = "sk-abc123..."` directly in your code and then share that code (on GitHub, with a friend, or when deploying online), **everyone can see your key**.

This is like writing your M-PESA PIN on the back of your phone — very dangerous! Someone could use your key and spend all your credits.

```python
# ❌ BAD — never do this when sharing or deploying code!
api_key = "sk-abc123xyz"    # anyone who sees this file can steal your key!
```

### The Solution — Streamlit Secrets

Streamlit has a built-in system called **`st.secrets`** for storing sensitive information like API keys safely.

It works like this:

**On your laptop** — create a file called `secrets.toml` inside a `.streamlit` folder:

```
Basic_Intro_to_GenAI/
├── .streamlit/
│   └── secrets.toml      ← your secret key goes here (never uploaded)
├── openai_chatbot_app.py
└── ollama_chatbot_app.py
```

Inside `secrets.toml`, write your key like this:
```toml
# .streamlit/secrets.toml
OPENAI_API_KEY = "sk-your-real-api-key-here"
```

**In your Python code** — instead of writing the key directly, you read it from `st.secrets`:

```python
# ✅ GOOD — the key is never written in the code itself!
api_key = st.secrets["OPENAI_API_KEY"]
client  = OpenAI(api_key=api_key)
```

Your code is now safe to share — it contains no key. The key only lives in `secrets.toml` on your laptop.

> ⚠️ **Always add `.streamlit/secrets.toml` to your `.gitignore` file** so it is never accidentally uploaded to GitHub!

````python
# Example: the updated openai_chatbot_app.py using st.secrets
# Compare this to the version that hardcodes the key — spot the difference!
# DO NOT run this cell in Jupyter — it only works inside a Streamlit .py file

import streamlit as st
from openai import OpenAI

# ✅ Read the API key from .streamlit/secrets.toml
# The key is stored safely outside your code
api_key = st.secrets["OPENAI_API_KEY"]

# Create the OpenAI client using the secret key
# Everything else in your chatbot stays exactly the same!
client = OpenAI(api_key=api_key)

st.title("🤖 Kamau AI")
# ... rest of your chatbot code continues as normal
````

**Streamlit Community Cloud** lets you host your chatbot online for **free**.

> 💡 **Important note:** Ollama runs only on your laptop, so only the **OpenAI version** can be deployed to the internet. When we deploy online, we use OpenAI (which runs in the cloud).

---

### Step-by-Step Deployment Guide

#### Step 1 — Prepare Your Files

Your project folder needs 3 files before deploying:

```
Basic_Intro_to_GenAI/
├── openai_chatbot_app.py   ← your chatbot (using st.secrets for the key)
├── requirements.txt        ← list of Python libraries needed
└── .gitignore              ← tells Git: "never upload my secrets.toml"
```

The next two code cells will create `requirements.txt` and `.gitignore` for you automatically!

---

#### Step 2 — Push Your Code to GitHub

1. Go to [github.com](https://github.com) and create a **free account** (if you don't have one)
2. Click the **"+"** button → **"New repository"**
3. Name it something like `kamau-chatbot` and click **Create repository**
4. Upload your files by dragging them into the repository page

> ✅ Double-check: `secrets.toml` should **NOT** appear in your GitHub repository

---

#### Step 3 — Deploy on Streamlit Community Cloud

1. Go to [share.streamlit.io](https://share.streamlit.io)
2. Click **"Sign in with GitHub"** — use the same GitHub account
3. Click **"New app"**
4. Fill in the form:
   - **Repository:** select your GitHub repo (e.g., `wanjiku/kamau-chatbot`)
   - **Branch:** `main`
   - **Main file path:** `openai_chatbot_app.py`
5. Click **"Advanced settings"** → **"Secrets"**
6. Paste your API key exactly like this:
   ```toml
   OPENAI_API_KEY = "sk-your-real-api-key-here"
   ```
7. Click **"Save"** then **"Deploy!"**

---

#### Step 4 — Share Your Link! 🎉

After about 1-2 minutes, Streamlit gives you a live public URL like:
```
https://wanjiku-kamau-chatbot.streamlit.app
```

Share this with your friends, teacher, or family — they can chat with your AI from any phone or computer!

### Summary: Local vs Deployed

| | On Your Laptop | Deployed Online |
|-|----------------|----------------|
| **Who can use it?** | Only you | Anyone with the link |
| **AI model** | Ollama (free, offline) | OpenAI GPT (needs API key) |
| **API key stored in** | `.streamlit/secrets.toml` | Streamlit Cloud secrets panel |
| **Cost** | Free | Free (Streamlit Community tier) |
| **Best for** | Building & testing | Sharing with the world |

````python
# Run this cell to automatically create requirements.txt
# This file tells Streamlit Cloud which libraries to install

requirements_content = """streamlit
openai
"""

with open("requirements.txt", "w") as f:
    f.write(requirements_content)

print("requirements.txt created!")
print("-" * 30)
print(requirements_content)
````

````python
# Run this cell to automatically create:
#   .gitignore          — protects your secrets from being uploaded to GitHub
#   .streamlit/secrets.toml  — template for your API key (fill in the real key!)

import os

# --- .gitignore ---
gitignore_content = """# Streamlit secrets — NEVER upload this to GitHub!
.streamlit/secrets.toml

# Python cache files
__pycache__/
*.pyc
"""

with open(".gitignore", "w") as f:
    f.write(gitignore_content)

print(".gitignore created!")

# --- .streamlit/secrets.toml ---
os.makedirs(".streamlit", exist_ok=True)   # create the folder if it doesn't exist

secrets_template = """# .streamlit/secrets.toml
# Replace the value below with your actual OpenAI API key.
# This file is listed in .gitignore — it will NOT be uploaded to GitHub.

OPENAI_API_KEY = "sk-your-api-key-here"
"""

with open(".streamlit/secrets.toml", "w") as f:
    f.write(secrets_template)

print(".streamlit/secrets.toml template created!")
print()
print("Next step: open .streamlit/secrets.toml and paste your real API key.")
````
