### Exercise 3A: What Does This Filter Detect?

Wanjiku is studying CNN filters at the University of Nairobi Computer Science department. She found these two filters. Run the code below, look at the output, then answer the questions.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# A simple test image — a bright square in the middle
test_image = np.zeros((10, 10))
test_image[3:7, 3:7] = 255  # Bright square in the centre

# Filter A — this is given to you
filter_A = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# Filter B — this is given to you
filter_B = np.array([
    [-1,  0,  1],
    [-1,  0,  1],
    [-1,  0,  1]
])

# Apply both filters
output_A = convolve2d(test_image, filter_A, mode='same')
output_B = convolve2d(test_image, filter_B, mode='same')

# Display results
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(test_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(output_A, cmap='gray')
axes[1].set_title('Filter A Output')
axes[1].axis('off')

axes[2].imshow(output_B, cmap='gray')
axes[2].set_title('Filter B Output')
axes[2].axis('off')

plt.tight_layout()
plt.show()
````
